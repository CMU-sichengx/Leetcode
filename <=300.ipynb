### 204 Count Primes
The sieve of Eratosthenes [erəˈtɒsθɪˌniːz]

````python
import math
class Solution:
    # Return the number of primes less than n
    def countPrimes(self, n: int) -> int:
        # Handle Degenerate Cases
        if n <= 2:
            return 0
        
        # Handle Normal Cases
        isPrime = [0, 0] + [1] * (n-2)
        
        for p in range(2, math.floor(math.sqrt(n))+1):
            if isPrime[p]:
                for mult in range(p*p, n, p):
                    isPrime[mult] = 0
        
        # print(isPrime)
        return sum(isPrime)
````

### 206 Reverse Linked List
- Common subroutine for linked list problems

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # three pointers pred, curr, succ
        pred = None
        curr = head

        while curr != None:
            succ = curr.next
            curr.next = pred
            pred = curr
            curr = succ
            
        return pred
        
````

````python
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        courseToPrereq = []
        for _ in range(numCourses):
            courseToPrereq.append([])
        
        for (a, b) in prerequisites:
            # add edge a -> b
            # adj[a].append(b)
            courseToPrereq[a].append(b)
        
        # returns if searching from 'course' has a cycle
        def has_cycle(course, visited, prevs):
            # Base Case
            if course in visited:
                return course in prevs
            
            # Recursive Case
            if course in prevs:
                return True
            visited.add(course)
            prevs.append(course)
            hasCycle = False
            # Recurse
            for node in courseToPrereq[course]:
                hasCycle = hasCycle or has_cycle(node, visited, prevs)
            prevs.pop()
            return hasCycle
        
        # iteratively apply dfs() on all the vertex
        visited = set()
        for course in range(numCourses):
            prevs = []
            if has_cycle(course, visited, prevs):
                return False
        return True
            
````

### 209 Minimum Size Subarray Sum
- Sliding window
- $!$ When not sure if the window can be restored in the 2nd step, add a check conditional in 3rd step
- keep track of the $\Sigma$ of the current array
- Two ways of writing step 2:
  - `Longest ...` Keep moving `begin` to right and stop once the window is restored
  - `Shortest ...` Keep moving `begin` to the right and stop right before the window is violated
  - Sometimes need to add a check conditional for if the window is restored at 3rd step

````python
def minSubArrayLen(self, target: int, nums: List[int]) -> int:
    begin, end = 0, 0
    width = math.inf
    total = 0
    
    while end < len(nums):
        # ------ 1. move end by one
        endNum = nums[end]
        total += endNum
        end += 1
        
        # ------ 2. move begin as long as the window is valid
        while True:
            beginNum = nums[begin]
            if total - beginNum >= target:
                total -= beginNum
                begin += 1
            else:
                break
        
        # ------ 3. Conditionally update the window length
        # print(begin, end, nums[begin : end], total)
        if total >= target:
            width = min(width, end-begin)
        
    return 0 if width == math.inf else width
````

### 213 House Robber II
- Reduce the problem to 2 old `house_robber` problems `[1:]` and `[:-1]`
- 大部分cycle都是转换成linear的问题

````python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        
        def houseRobberLinear(nums):
            # handle edge cases
            if len(nums) == 0:
                return 0
            if len(nums) == 1:
                return nums[0]
            # handle normal cases
            n = len(nums)
            No = 0
            Yes = nums[0]
            
            for i in range(2, n+1):
                newNo = max(Yes, No)
                newYes = nums[i-1] + No
                No = newNo
                Yes = newYes
            return max(No, Yes)
            
        # Handle edge cases
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]
        if len(nums) == 2:
            return max(nums[0], nums[1])
        # Especially considering house[0], there are only two options, consider
        # them all
        # Option 1: rob the first house, skip the last house
        money1 = houseRobberLinear(nums[1:])
        # Option 2: skip the first house, rob the last house
        money2 = houseRobberLinear(nums[:len(nums)-1])
        return max(money1, money2)
````

### 216 Combination Sum III
 - Backtrack (naive version)

````python
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        combs = []
        
        def backtrack(cursor, comb_so_far, sum_so_far):
            if len(comb_so_far) == k:
                if sum_so_far == n:
                    combs.append(comb_so_far.copy())
                return
            if sum_so_far > n:
                return
            
            for new_cursor in range(cursor, 10):
                comb_so_far.append(new_cursor)
                backtrack(new_cursor+1, comb_so_far, sum_so_far+new_cursor)
                comb_so_far.pop()
        
        backtrack(1, [], 0)
        return combs
````

### 222 Count Complete Tree Nodes
 - Binary search tree inside binary search tree costs $O(log^2(n))$ time
 - Binary representation
 - What we know from `complete` tree -> 
    - binary search range is `how many nodes there are at the bottom level`
    - the path from root to each leaf index can be inferred from the binary vector of its label $2^d$ to $2^{d+1}-1$ 
    - Each query takes $log(n)$ time, and the binary search makes at most $log(n)$ query.

````python
def getBinary(self, n):
    res = ""
    while n > 0:
        if n % 2 == 0:
            res = '0' + res
        else:
            res = '1' + res
        n = n // 2
    return res
    
def countNodes(self, root: Optional[TreeNode]) -> int:
    # handle degenerate case
    if root is None: return 0
    
    # handle normal case
    # ----- Compute the number of complete levels
    d = 0
    cursor = root
    while cursor:
        cursor = cursor.right
        d += 1
    
    
    def leafExist(i):
        cursor = root
        binString = self.getBinary(i)
        for c in binString[1:]:
            if c == '0': 
                if not cursor.left: return False
                cursor = cursor.left
            else: 
                if not cursor.right: return False
                cursor = cursor.right
        return True
    
    # ----- now we know that the last level node ranges from 
    # ----- labelling starting from one
    # 2^d  . . . . . 2^(d+1)-1               
    # ----- perform binary search on this range
    lo, hi = int(math.pow(2, d)), int(math.pow(2, d+1))
    
    while lo < hi:
        mid = (lo + hi) // 2
        if leafExist(mid):
            lo = mid + 1
        else:
            hi = mid

    return lo - 1
````

### 236 Lowest Common Ancestor of a Binary Tree
 - Use property: the stack returns from the base case call to the original function call
 - `if n == p or n == q: return n` $\implies$ a node is found
 - Execution flow: whenever a node `p` or `q` is found, that node is returned:
   - case 1:  LCA $\in$ `{p, q}` : By property of the stack, whoever is higher in the tree will be returned later in the function call stack $\implies$ whoever between `p` and `q` is higher will overwrite the other one $\implies$ the LCA is correctly returned
   - case 2: LCA $\notin$ `{p, q}` This will be detecetd for the first time by `if left_flag and right_flag: return n`. From then on,`n` is always to its caller until the original function call/

````python
def LCA(self, n, p, q):
    if n is None: return None
    if n == p or n == q: return n

    left_flag = self.LCA(n.left, p, q)
    right_flag = self.LCA(n.right, p, q)
    if left_flag and right_flag: return n
    if not left_flag and not right_flag: return None
    else: return left_flag if left_flag else right_flag
````

### 240 Search a 2D Matrix II
 - $O(m + n)$ Two pointer Method 
 - $\textit{Space Elimination}$

````python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m = len(matrix)
    n = len(matrix[0])
    
    r, c = m-1, 0 # the bottom left corner
    
    # Invariant: the target, if exists, must lie within
    # (Note all boundaries inclusive)
    #   _______ (0,n-1)
    #  |           |
    #  |           |
    #  |           |
    #  |           |
    # (r,c)________|
    
    while 0 <= r < m and 0 <= c < n:
        if matrix[r][c] == target:
            return True
        elif matrix[r][c] < target:
            c += 1
        else:
            r -= 1
    return False
````

### 259 3Sum Smaller
 - Two pointer `nums.sort()` and `lo = 0` and `hi = n-1`
    - Useful for: finding number of combinations of all `(i, j)` that satisfy a certain property
        - 本题: how many `(i, j)`s that satisfy the property that `nums[i] + nums[j] < target`
    - pseudocode    
    ```
    Init res
    Init lo, hi
    while lo < hi:
        curSum = A[lo] + A[hi]
        Update res
        casing by comparing curSum and target
    ```

````python
def twoSumSmaller(self, start, nums, target):
    lo = start
    hi = len(nums) - 1
    
    count = 0
    while lo < hi:
        curSum = nums[lo] + nums[hi]
        if curSum >= target:
            hi -= 1
        elif curSum < target:
            # [lo, hi] is now a valid
            # ---> [lo, lo+1], [lo, lo+2], ..., [lo, hi] are all valid
            count += hi - lo
            lo += 1
    return count
            
def threeSumSmaller(self, nums: List[int], target: int) -> int:
    nums.sort()
    count = 0
    
    for i, num1 in enumerate(nums):
        count += self.twoSumSmaller(i+1, nums, target-num1)
    return count
````

### 261 Graph Valid Tree
- 流程 construct adj list $\implies$ check if all vertices are reachable using `coverGraph` $\implies$ check for no cycles by comparing `len(edges)` and `n-1`

````python
class Solution(object):
    def validTree(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: bool
        checking for connectedness and non-cycle
        """
        
        nodeToNbors = []
        for i in range(n):
            nodeToNbors.append([])
        
        for (a, b) in edges:
            nodeToNbors[a].append(b)
            nodeToNbors[b].append(a)
        
        # Check (1) the graph is connected
        #       (2) |V| = |E| + 1
        
        # Now we are sure that the graph does not contain cycles
        # starting from any node should be able to cover the graph
        reachable = [False] * n
        visited = set()
        
        def coverGraph(node):
            # Base Case
            if node in visited:
                return
            
            # Recursive Case
            reachable[node] = True
            visited.add(node)
            for nbor in nodeToNbors[node]:
                coverGraph(nbor)
            return
        
        coverGraph(0)
        for b in reachable:
            if not b:
                return False
        return len(edges) == n-1
````

### 270 Closest Binary Search Tree Value
 - go left if target is smaller than current root value, and go right otherwise. Keep track of the closest to target value at each step
 - $BinarySeaerch$
    - Search space `all the nodes in the tree`. At each step, eliminate roughly half by going either left or the right
    - If `target` greater than cur val, then right subtree eliminated
    - If `target` smaller than cur val, then left subtree eliminated

````python
def closestValue(self, root: Optional[TreeNode], target: float) -> int:
    # --- return the one closer one betwen {val1, val2} and target
    def getCloser(val1, val2):
        diff1 = abs(val1 - target)
        diff2 = abs(val2 - target)
        return val1 if diff1 < diff2 else val2
    
    closest = -math.inf
    while root:
        closest = getCloser(closest, root.val)
        if target == root.val: break
        elif target < root.val:
            root = root.left
        else:
            root = root.right
    return closest
````

### 271 Encode and Decode Strings
- str(100) = '100'

````python
class Codec:
    def encode(self, strs):
        """Encodes a list of strings to a single string.
        
        :type strs: List[str]
        :rtype: str
        """
        msg = ""
        lengths = ""
        for s in strs:
            msg += s
            length = str(len(s))
            if len(s) < 10:
                length = "00" + length
            if 10 <= len(s) < 100:
                length = "0" + length
            lengths += length
        D = len(lengths)
        M = str(D) + "#" + lengths + msg
        return M
        

    def decode(self, s):
        """Decodes a single string to a list of strings.
        
        :type s: str
        :rtype: List[str]
        """
        p = 0 # the pointer used to read the string
        
        # Parse length of lengths part
        while s[p] != '#':
            p += 1
        length = s[0:p]
        D = int(length)
        lengths = s[p+1 : p+1+D]
        words = s[p+1+D : ]
        strs = []

        length_reader = 0
        word_reader = 0
        for _ in range(D // 3):
            length_str = lengths[length_reader : length_reader+3]
            length_int = int(length_str)
            length_reader += 3
            word = words[word_reader : word_reader+length_int]
            word_reader += length_int
            strs.append(word)
        return strs
            
# Your Codec object will be instantiated and called as such:
# codec = Codec()
# codec.decode(codec.encode(strs))
````

### 277 Find the Celebrity
 - Pattern: 
   - Query mode: each `query(x, y)` can eliminate either `x` or `y`
   - Want to query in a way to avoid no redundant elimination $\implies$ only need `n-1` queries

````python
def findCelebrity(self, n: int) -> int:
    candidate = 0
    
    # eliminate n-1 people
    for i in range(1, n):
        # invariant1: candidate < i at the beginning of each iteration
        # invariant2: possible range is {candidate} U {i, i+1, ..., n-1)
        # invariant3: candidate is the only one possible out of {0, 1, ..., i-1}
        if knows(candidate, i):
            candidate = i
            
    # varify the remaining one
    for i in range(n):
        if i == candidate: continue
        if knows(candidate, i):
            return -1
        if not knows(i, candidate):
            return -1
    return candidate
````

### 283 Move Zeroes

````python
def moveZeroes(self, nums: List[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    wr = 0
    rd = 0
    while rd < len(nums):
        if nums[rd] == 0:
            rd += 1
        else:
            nums[wr] = nums[rd]
            rd += 1
            wr += 1
            
    while wr < len(nums):
        nums[wr] = 0
        wr += 1
````

### 289 Game of Life
 - Store the information about the new state and the old state at the same time using no more bits
 - New encoding system in the comment

````python
class Solution:
    def gameOfLife(self, board: List[List[int]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        rows = len(board)
        cols = len(board[0])

        # new encoding of the state that is compatible with the old 
        # encoding system
        # 'OLD'
        # die -> ?       0
        # live -> ?      1
        
        # 'NEW'
        # live -> live   1
        # live -> die    -1
        # die -> live    2
        # die -> die     0
        
        # criteria
        def used_to_live(r, c): return abs(board[r][c]) == 1 
        def now_live(r, c): return board[r][c] > 0
        def within(r, c): return 0 <= r < rows and 0 <= c < cols
        
        for r in range(rows):
            for c in range(cols):
                liveNbors = 0
                # count the neighboring cells of this (r, c) cell
                for dRow in [-1, 0, 1]:
                    for dCol in [-1, 0, 1]:
                        if dRow == 0 and dCol == 0: continue
                        newR = r + dRow
                        newC = c + dCol
                        if within(newR, newC) and used_to_live(newR, newC):
                            liveNbors += 1
                # assign new state 
                if board[r][c] == 1:
                    if (liveNbors < 2 or liveNbors > 3):
                        board[r][c] = -1 # live -> die
                    else:
                        board[r][c] = 1 # live -> live
                else:
                    if liveNbors == 3:
                        board[r][c] = 2 # die -> live
                    else:
                        board[r][c] = 0 # die -> die
        
        # restore to the old encoding system
        for r in range(rows):
            for c in range(cols):
                if now_live(r, c):
                    board[r][c] = 1
                else:
                    board[r][c] = 0
````

### 300 Longest Increasing Subsequence
- Define `DP[i]` := max length of LIS that ends at index i
- Recurrence $DP[i] = \max_{j < i: arr[j] < arr[i]} 1+DP[j]$

````python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Define DP[i] := max length of LIS that ends at index i
        n = len(nums)
        DP = [1] * n
        
        # Base Case
        DP[0] = 1
        
        # Recursive Case
        for i in range(1, n):
            curr = nums[i]
            for j in range(i):
                if nums[j] < curr:
                    DP[i] = max(DP[i], 1+DP[j])

        # take the max of all ending indices
        ans = 0
        for i in range(n):
            ans = max(ans, DP[i])
        return ans
````
