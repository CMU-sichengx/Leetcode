### 202 Happy Number

````python
class Solution:
    def isHappy(self, n: int) -> bool:        
        def go_next(v):
            next_v = 0
            for d_str in str(v):
                d_int = int(d_str)
                next_v += d_int * d_int
            return next_v
        
        turtle = n
        rabbit = go_next(n)
        if turtle == 1 or rabbit == 1:
            return True
        
        # Traverse the implicit linked list
        while go_next(rabbit) != 1 and go_next(go_next(rabbit)) != 1:
            if rabbit == turtle:
                return False
            rabbit = go_next(go_next(rabbit))
            turtle = go_next(turtle)
        return True
````

### 203 Remove Linked List Elements
 - Easy: keep track of `prev` and `curr` and update correspondingly
 - Use a dummy header in case the first node needs to be removed

````python
def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
    dummy = ListNode(0, head)
    prev, curr = dummy, head
    while curr:
        if curr.val == val: 
            prev.next = curr.next
            curr = curr.next
        else:
            prev = curr
            curr = curr.next
    
    return dummy.next
````

### 204 Count Primes
The sieve of Eratosthenes [erəˈtɒsθɪˌniːz]

````python
import math
class Solution:
    # Return the number of primes less than n
    def countPrimes(self, n: int) -> int:
        # Handle Degenerate Cases
        if n <= 2:
            return 0
        
        # Handle Normal Cases
        isPrime = [0, 0] + [1] * (n-2)
        
        for p in range(2, math.floor(math.sqrt(n))+1):
            if isPrime[p]:
                for mult in range(p*p, n, p):
                    isPrime[mult] = 0
        
        # print(isPrime)
        return sum(isPrime)
````

### 205 Isomorphic Strings
 - Method 1: Keep track of established mappings (track `domain` via `matched` and track `codomain` via `codomain`)
 - Method 2: Encode a string: encode each character as the `index of its first occurance in ths string`

````python
def isIsomorphic(self, s: str, t: str) -> bool:
    matched = dict() # used to store all the established mappings
    codomain = set() # the codomain of the established mapping
    if len(s) != len(t): return False
    
    for c1, c2 in zip(s, t):
        if c1 not in matched:
            if c2 not in codomain:
                matched[c1] = c2
                codomain.add(c2)
            else: # c2 is already occupied
                return False
        else:
            if matched[c1] != c2: return False
    return True
    
def isIsomorphic(self, s: str, t: str) -> bool:
        def encode(s):
            letter2idx = dict()
            encode = []
            for i, c in enumerate(s):
                if c not in letter2idx:
                    letter2idx[c] = str(i)
                encode.append(letter2idx[c])
            return ' '.join(encode)
        
        # Driver
        return encode(s) == encode(t)
        
````

### 209 Minimum Size Subarray Sum
- Sliding window
- $!$ When not sure if the window can be restored in the 2nd step, add a check conditional in 3rd step
- keep track of the $\Sigma$ of the current array
- Two ways of writing step 2:
  - `Longest ...` Keep moving `begin` to right and stop once the window is restored
  - `Shortest ...` Keep moving `begin` to the right and stop right before the window is violated
  - Sometimes need to add a check conditional for if the window is restored at 3rd step

````python
def minSubArrayLen(self, target: int, nums: List[int]) -> int:
    begin, end = 0, 0
    width = math.inf
    total = 0
    
    while end < len(nums):
        # ------ 1. move end by one
        endNum = nums[end]
        total += endNum
        end += 1
        
        # ------ 2. move begin as long as the window is valid
        while True:
            beginNum = nums[begin]
            if total - beginNum >= target:
                total -= beginNum
                begin += 1
            else:
                break
        
        # ------ 3. Conditionally update the window length
        # print(begin, end, nums[begin : end], total)
        if total >= target:
            width = min(width, end-begin)
        
    return 0 if width == math.inf else width
````

### 213 House Robber II
- Reduce the problem to 2 old `house_robber` problems `[1:]` and `[:-1]`
- 大部分cycle都是转换成linear的问题

````python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        def houseRobberLinear(nums):
            # handle edge cases
            if len(nums) == 0:
                return 0
            if len(nums) == 1:
                return nums[0]
            # handle normal cases
            n = len(nums)
            No = 0
            Yes = nums[0]
            
            for i in range(1, n):
                newNo = max(Yes, No)
                newYes = nums[i] + No
                No = newNo
                Yes = newYes
            return max(No, Yes)
            
        # Handle edge cases
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]
        if len(nums) == 2:
            return max(nums[0], nums[1])
        # Especially considering house[0], there are only two options, consider
        # them all
        # Option 1: rob the first house, skip the last house
        money1 = houseRobberLinear(nums[1:])
        # Option 2: skip the first house, rob the last house
        money2 = houseRobberLinear(nums[:len(nums)-1])
        return max(money1, money2)
````

### 216 Combination Sum III
 - Backtrack (naive version)

````python
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        combs = []
        
        def backtrack(cursor, comb_so_far, sum_so_far):
            if len(comb_so_far) == k:
                if sum_so_far == n:
                    combs.append(comb_so_far.copy())
                return
            if sum_so_far > n:
                return
            
            for new_cursor in range(cursor, 10):
                comb_so_far.append(new_cursor)
                backtrack(new_cursor+1, comb_so_far, sum_so_far+new_cursor)
                comb_so_far.pop()
        
        backtrack(1, [], 0)
        return combs
````

### 221 Maximal Square
 - Dynamic Programming 
 - Recursion $$DP[r,c] = min(DP[r-1,c], DP[r, c-1], DP[r-1,c-1]) + 1$$
 - Keep track of only two rows for space efficiency $O(n)$

### 222 Count Complete Tree Nodes
 - Binary search tree inside binary search tree costs $O(log^2(n))$ time
 - Binary representation
 - What we know from `complete` tree -> 
    - binary search range is `how many nodes there are at the bottom level`
    - the path from root to each leaf index can be inferred from the binary vector of its label $2^d$ to $2^{d+1}-1$ 
    - Each query takes $log(n)$ time, and the binary search makes at most $log(n)$ query.

````python
def getBinary(self, n):
    res = ""
    while n > 0:
        if n % 2 == 0:
            res = '0' + res
        else:
            res = '1' + res
        n = n // 2
    return res
    
def countNodes(self, root: Optional[TreeNode]) -> int:
    # handle degenerate case
    if root is None: return 0
    
    # handle normal case
    # ----- Compute the number of complete levels
    d = 0
    cursor = root
    while cursor:
        cursor = cursor.right
        d += 1
    
    
    def leafExist(i):
        cursor = root
        binString = self.getBinary(i)
        for c in binString[1:]:
            if c == '0': 
                if not cursor.left: return False
                cursor = cursor.left
            else: 
                if not cursor.right: return False
                cursor = cursor.right
        return True
    
    # ----- now we know that the last level node ranges from 
    # ----- labelling starting from one
    # 2^d  . . . . . 2^(d+1)-1               
    # ----- perform binary search on this range
    lo, hi = int(math.pow(2, d)), int(math.pow(2, d+1))
    
    while lo < hi:
        mid = (lo + hi) // 2
        if leafExist(mid):
            lo = mid + 1
        else:
            hi = mid

    return lo - 1
````

### 223 Rectangle Area
 - Geometry, compute area 题目
 - Technique 1: Compute the overlap area and then compute `s1 + s2 - overlap`
 - Technique 2: Break down 2-dimensional problem to two 1-dimensional problem

````python
def computeArea(self, ax1: int, ay1: int, ax2: int, ay2: int, bx1: int, by1: int, bx2: int, by2: int) -> int:
    L1 = ax1
    R1 = ax2
    U1 = ay2
    D1 = ay1
    
    L2 = bx1
    R2 = bx2
    U2 = by2
    D2 = by1
    
    # 1-- Compute the overlapping area
    # considering from x axis (L1, R1) and (L2, R2)
    w = min(R1, R2) - max(L1, L2)
    w = max(0, w)
    # considering from y axis (D1, U1) and (D2, U2)
    h = min(U1, U2) - max(D1, D2)
    h = max(0, h)
    overlap = w * h
    
    s1 = (R1-L1) * (U1-D1)
    s2 = (R2-L2) * (U2-D2)
    return s1 + s2 - overlap
````

### 226 Invert Binary Tree
 - Approach One: `recursively invert left subtree` + `recursively invert the right subtree` + `swap left and right`
 - Approach Two: `Using a deque to be iterative`
 - Approach Three: `level order traversal`

````python
class Solution(object):
    def invertTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        # Base Case
        if root is None:
            return None
        # Recursive Case
        leftInverted = self.invertTree(root.left)
        rightInverted = self.invertTree(root.right)
        root.left = rightInverted
        root.right = leftInverted
        return root
        
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        Q = deque([root])
        while Q:
            node = Q.popleft()
            node.left, node.right = node.right, node.left
            
            if node.left: Q.append(node.left)
            if node.right: Q.append(node.right)
        return root
````

### 228 Summary Ranges
 - Easy
 - Iterate through the array
 - Each time constructing one `range`, and append it onto the `ranges`
 - To construct each `range` per iteration, 
    - `start` is the starting index of the current `range`
    - `end` is the first earlies index such that `start` to `end` will not be a range
    - 1. `[start ... end-1]` both inclusive is the current range
    - 2. `[end]` is the new `start'`

````python
def summaryRanges(self, nums: List[int]) -> List[str]:
    # handle degenerate case
    if len(nums) == 0: return []
    if len(nums) == 1: return [str(nums[0])] # this line is not needed. Only for an erronous test 
    
    # handle normal cases
    ranges = []
    start, end = 0, 0
    while end < len(nums):
        # 1-- move end to right as long as number is consecutive
        while end < len(nums):
            if end == start: end += 1
            elif nums[end-1] == nums[end] or nums[end-1] + 1 == nums[end]: end += 1
            else: break
        # 2-- now [start, end) is the new consecutive range
        thisRange = str(nums[start]) if nums[start] == nums[end-1] else \
                    str(nums[start]) + '->' + str(nums[end-1])
        ranges.append(thisRange)
        # 3-- go to fetch the next range
        start = end
    
    return ranges
        
````

### 229 Majority Element II
 - $\textit{Boyer-Moore Voting Algorithm}$
 - Keep track of `major1, major2, est1, est2` 
 - Proof: For each incoming element `num`:
    - if neither current `major1` or `major2` is the actual major element: the incoming `num` will be used to offset false majors.
    - If either current `major1` or `major2` is the actual major element: the incoming `num` will be used to offset its `est` by strictly less that $\frac{n}{3}$ times. However, we know that `est1` and `est2` will be incremented by strictly more than $\frac{n}{3}$ times. So, its `est` will be positive. The correct major will be retained in the `major1` and `major2`
 - Python count buitlin 
   ```
   # create a list
   numbers = [2, 3, 5, 2, 11, 2, 7]
   # check the count of 2
   count = numbers.count(2)
   ```

````python
class Solution:
def majorityElement(self, nums: List[int]) -> List[int]:
    # there are at most 2 elements as majority
    major1, major2 = None, None
    est1, est2 = 0, 0

    for num in nums:
        # if num is already major1 or major2
        if num == major1: est1 += 1
        elif num == major2: est2 += 1

        # if there is an empty spot
        elif major1 is None: 
            major1 = num
            est1 = 1
        elif major2 is None: 
            major2 = num
            est2 = 1

        # if there is no empty stop, eliminate both current major1 and major2 by one
        else:
            est1 -= 1
            if est1 == 0: major1 = None
            est2 -= 1
            if est2 == 0: major2 = None
    #print(major1, major2)
    
    # traverse the array again to check if it is a real majority
    res = []
    if nums.count(major1) > floor(len(nums) / 3): res.append(major1)
    if nums.count(major2) > floor(len(nums) / 3): res.append(major2)
    return res
````

````python
def __init__(self):
    self.instack = []
    self.outstack = []

def push(self, x: int) -> None:
    self.instack.append(x)

def _transfer(self):
    while self.instack:
        ele = self.instack.pop()
        self.outstack.append(ele)
        
def pop(self) -> int:
    if len(self.outstack) == 0: self._transfer()
    return self.outstack.pop()

def peek(self) -> int:
    if len(self.outstack) == 0: self._transfer()
    return self.outstack[-1]

def empty(self) -> bool:
    return len(self.instack) + len(self.outstack) == 0
````

### 235 Lowest Common Ancestor of a Binary Search Tree
 - The lowest common ancestor algo

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    '''
    p and q are guaranteed to be two distinct, non-null nodes in the tree,
    also there is guaranteed to be a solution
    '''
    def LCA_BST(self, root, p, q):
        # Base Case
        if p.val <= root.val <= q.val:
            return root
        # Recursive Case
        if p.val < root.val:
            return self.LCA(root.left, p, q)
        else:
            return self.LCA(root.right, p, q)
        
        
        
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if p.val > q.val: p, q = q, p
        # p is the smaller node and q is the larger node
        return self.LCA_BST(root, p, q)
        
````

### 236 Lowest Common Ancestor of a Binary Tree
 - Use property: the stack returns from the base case call to the original function call
 - `if n == p or n == q: return n` $\implies$ a node is found
 - Execution flow: whenever a node `p` or `q` is found, that node is returned:
   - case 1:  LCA $\in$ `{p, q}` : By property of the stack, whoever is higher in the tree will be returned later in the function call stack $\implies$ whoever between `p` and `q` is higher will overwrite the other one $\implies$ the LCA is correctly returned
   - case 2: LCA $\notin$ `{p, q}` This will be detecetd for the first time by `if left_flag and right_flag: return n`. From then on,`n` is always to its caller until the original function call/

````python
def LCA(self, n, p, q):
    if n is None: return None
    if n == p or n == q: return n

    left_flag = self.LCA(n.left, p, q)
    right_flag = self.LCA(n.right, p, q)
    if left_flag and right_flag: return n
    if not left_flag and not right_flag: return None
    else: return left_flag if left_flag else right_flag
````

### 239 Sliding Window Maximum
 - Monotonic Queue (The queue is always arranged in decreasing order)
    - Sorted descending order
 - See implementation
 

````python
from collections import deque

class monotonicQueue:
    def __init__(self):
        self.data = deque([])
        
    def popLeft(self):
        if len(self.data) > 0:
            self.data.popleft()
    
    def appendRight(self, num):
        while self.data and self.data[-1] < num:
            self.data.pop()
        self.data.append(num)
    def getMax(self):
        return self.data[0]

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        n = len(nums)
        windowMaxs = []
        Q = monotonicQueue()
        
        for i in range(n):
            Q.appendRight(nums[i])
            
            if i + 1 >= k: # When >= k numbers present in the queue
                windowMaxs.append(Q.getMax())
                firstInWindow = nums[i-k+1] # leftmost element of the window after including [i]
                if firstInWindow == Q.getMax(): Q.popLeft() # this number can no longer be used

        return windowMaxs
````

### 240 Search a 2D Matrix II
 - $O(m + n)$ Two pointer Method 
 - $\textit{Space Elimination}$

````python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m = len(matrix)
    n = len(matrix[0])
    
    r, c = m-1, 0 # the bottom left corner
    
    # Invariant: the target, if exists, must lie within
    # (Note all boundaries inclusive)
    #   _______ (0,n-1)
    #  |           |
    #  |           |
    #  |           |
    #  |           |
    # (r,c)________|
    
    while 0 <= r < m and 0 <= c < n:
        if matrix[r][c] == target:
            return True
        elif matrix[r][c] < target:
            c += 1
        else:
            r -= 1
    return False
````

### 241 Different Ways to Add Parentheses
 - For mathematical operations in the form of `X o Y` -> think about optimal binary search tree DP structure
 - DP base case `s[i, j) is all digits` -> a singleton list containing just the integer istself
 - DP recursive case: Loop through the string. Set all operators as the division point. Recurse on the left part and the right part

````python
def diffWaysToCompute(self, expression: str) -> List[int]:
    cache = dict()
    
    Ops = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '*': lambda x, y: x * y
    }
    
    # calculate the number of different results of s[i:j)
    def dp(i, j):
        # Base case
        if expression[i:j].isdigit(): 
            cache[(i, j)] = [int(expression[i:j])] # singleton possibility
            return cache[(i, j)]
        
        # Recursive case
        res = []
        for k in range(i, j):
            if expression[k] in '+-*/':
                lefts = dp(i, k)
                rights = dp(k+1, j)
                for l in lefts:
                    for r in rights:
                        res.append(Ops[expression[k]](l, r))
        cache[(i, j)] = res
        return cache[(i, j)]
    
    # Driver
    return dp(0, len(expression))
````

### 243 Shortest Word Distance
 - One pass $O(n)$
 - Keep track of `i1, i2` pointing to the most recent index of previous occuerance of `word1` and `word2`
 - Update the $\min{|i_1 - i_2|}$ along each iteration

````python
def shortestDistance(self, wordsDict: List[str], word1: str, word2: str) -> int:
    # handle degenerate case
    if word1 == word2: return 0
    
    # handle normal cases
    i1, i2 = -1, -1
    minDist = math.inf
    
    # inv: i1, i2 keeps track of the most recent index of word1 and word2
    for i, word in enumerate(wordsDict):
        if word == word1: i1 = i
        if word == word2: i2 = i
        if i1 > -1 and i2 > -1: minDist = min(minDist, abs(i1-i2))
        
    return minDist
````

### 244 Shortest Word Distance II
 - Keep track of a dict mapping each `word` to its `list of indices of occurances`
 - Every time using linear time to traverse the two list of indices to obain $\min{|i_1 - i_2|}$

````python
def __init__(self, wordsDict: List[str]):
    self.locs = dict() # mapping word to its indices of occurance
    for i, word in enumerate(wordsDict):
        if word not in self.locs:
            self.locs[word] = [i]
        else:
            self.locs[word].append(i)
            
def shortest(self, word1: str, word2: str) -> int:
    indices1 = self.locs[word1]
    indices2 = self.locs[word2]
    r1 , r2 = 0, 0
    minDist = math.inf
    while r1 < len(indices1) and r2 < len(indices2):
        i1 = indices1[r1]
        i2 = indices2[r2]
        minDist = min(minDist, abs(i1-i2))
        if i1 == i2: return 0
        elif i1 < i2: r1 += 1
        else: r2 += 1
    return minDist
````

### 246 Strobogrammatic Number
 - Horizontal Flip : `reverse`
 - Vertical Flip : 
            ```
            flip = {
                        '0' : '0',
                        '1' : '1',
                        '6' : '9',
                        '8' : '8',
                        '9' : '6'
                    }
            ```

````python
def isStrobogrammatic(self, num: str) -> bool:
    # scan for 0, 1, 6, 8, 9
    for d in num:
        if d not in ('0', '1', '6', '8', '9'): return False
    
    # now all the numbers are 1, 6, 8, 9
    flip = {
        '0' : '0',
        '1' : '1',
        '6' : '9',
        '8' : '8',
        '9' : '6'
    }
    
    digits = list(map(lambda x : flip[x], reversed(list(num))))
    revNum = ''.join(digits)
    return num == revNum
````

### 250 Count Univalue Subtrees
 - $Intuition$ Think recursion in term of the tree structure. The execution begins from the bottom leaves and goes upwards in the tree. The conuter stops incrementing once the value becomes different.

````python
def countUnivalSubtrees(self, root: Optional[TreeNode]) -> int:
    # handle degeneraet case
    if root is None: return 0
    
    # handl normal cases
    # helper function returns (cnt, unival)
    def helper(root):
        # base case 1
        if root is None: return (0, -math.inf)
        
        # base case 2
        if root.left is None and root.right is None: return (1, root.val)
        
        # recursive case
        (cntLeft, valLeft) = helper(root.left)
        (cntRight, valRight) = helper(root.right)
        
        # 1-- both left and right children exist
        if root.left and root.right:
            if root.val == valLeft and root.val == valRight:
                cnt = 1 + cntLeft + cntRight
                unival = root.val
                return (cnt, unival)
            else:
                cnt = cntLeft + cntRight
                unival = -math.inf
                return (cnt, unival)
        
        # 2-- only has only one chlid
        else:
            onlyVal = valLeft if root.left else valRight
            onlyCnt = cntLeft if root.left else cntRight
            if root.val == onlyVal:
                cnt = 1 + onlyCnt
                unival = root.val
                return (cnt, unival)
            else: 
                cnt = onlyCnt
                unival = -math.inf
                return (cnt, unival)
    # Driver
    (cnt, unival) = helper(root)
    return cnt
````

### 254 Factor Combinations
 - Sol 1: Backtrack TLE

````python
def getFactors(self, n: int) -> List[List[int]]:
    # handle degenerate cases
    if n < 2: return []
    
    # handle normal cases
    facs = []
    
    def backtrack(curFac, curProd):
        # base case
        if curProd > n: return
        if curProd == n:
            facs.append(curFac.copy())
            return 
        
        # recursive case
        lastFactor = 2 if len(curFac) == 0 else curFac[-1]
        for newFactor in range(lastFactor, n//lastFactor+1):
            if n % newFactor == 0:
                curFac.append(newFactor)
                backtrack(curFac, curProd * newFactor)
                curFac.pop()
    
    # Driver
    backtrack([], 1)
    return facs
````

### 259 3Sum Smaller
 - Two pointer `nums.sort()` and `lo = 0` and `hi = n-1`
    - Useful for: finding number of combinations of all `(i, j)` that satisfy a certain property
        - 本题: how many `(i, j)`s that satisfy the property that `nums[i] + nums[j] < target`
    - pseudocode    
    ```
    Init res
    Init lo, hi
    while lo < hi:
        curSum = A[lo] + A[hi]
        Update res
        casing by comparing curSum and target
    ```

````python
def twoSumSmaller(self, start, nums, target):
    lo = start
    hi = len(nums) - 1
    
    count = 0
    while lo < hi:
        curSum = nums[lo] + nums[hi]
        if curSum >= target:
            hi -= 1
        elif curSum < target:
            # [lo, hi] is now a valid
            # ---> [lo, lo+1], [lo, lo+2], ..., [lo, hi] are all valid
            count += hi - lo
            lo += 1
    return count
            
def threeSumSmaller(self, nums: List[int], target: int) -> int:
    nums.sort()
    count = 0
    
    for i, num1 in enumerate(nums):
        count += self.twoSumSmaller(i+1, nums, target-num1)
    return count
````

### 261 Graph Valid Tree
- 流程 construct adj list $\implies$ check if all vertices are reachable using `coverGraph` $\implies$ check for no cycles by comparing `len(edges)` and `n-1`

````python
class Solution(object):
    def validTree(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: bool
        checking for connectedness and non-cycle
        """
        
        nodeToNbors = []
        for i in range(n):
            nodeToNbors.append([])
        
        for (a, b) in edges:
            nodeToNbors[a].append(b)
            nodeToNbors[b].append(a)
        
        # Check (1) the graph is connected
        #       (2) |V| = |E| + 1
        
        # Now we are sure that the graph does not contain cycles
        # starting from any node should be able to cover the graph
        reachable = [False] * n
        visited = set()
        
        def coverGraph(node):
            # Base Case
            if node in visited:
                return
            
            # Recursive Case
            reachable[node] = True
            visited.add(node)
            for nbor in nodeToNbors[node]:
                coverGraph(nbor)
            return
        
        coverGraph(0)
        for b in reachable:
            if not b:
                return False
        return len(edges) == n-1
````

### 265 Paint House II
 - The `remove()` method removes the first matching element (which is passed as an argument) from the list.
 - Space Efficient Solution: Only needs to keep track of the smallest and second smallest costs so far.
 - See sol 2  
    - Time efficient: we don't need to look at the entire previous row for every cell. When we're calculating the values for the second row, we're adding the minimum from the first row onto them or the second smallest only if the color-adjacency rule is violated
    - keep track of the 2 smallest values we've seen so far

````python
def minCostII(self, costs: List[List[int]]) -> int:
    n = len(costs)
    k = len(costs[0])
    
    # Base case: calculating the min cost of painting the first 1 building
    DP = costs[0].copy() # DP[i] = cost of priniting the previous building using i
    
    for r in range(2, n+1):
        
        # calcaulte the min cost of painting the first r buildings 
        newDP = [0] * k
        for color in range(k):
            # extract the prev cost with last color not equal to current color
            prevCost = math.inf
            for prevColor in range(k):
                if color == prevColor: continue
                prevCost = min(prevCost, DP[prevColor])       
            newDP[color] = costs[r-1][color] + prevCost
        DP = newDP
        
    return min(DP)

# O(nk) time and O(1) space
def twoMins(self, row):
    min1, i1 = row[0], 0
    min2, i2 = row[1], 1
    
    # Swap if needed
    if min1 > min2:
        min1, i1, min2, i2 = min2, i2, min1, i1
    
    for i, num in enumerate(row[2:]):
        i = i + 2
        if num < min1:
            min2, i2 = min1, i1
            min1, i1 = num, i
        elif num < min2:
            min2, i2 = num, i
    
    return (min1, i1, min2, i2)
            

def minCostII(self, costs: List[List[int]]) -> int:
    n = len(costs)
    k = len(costs[0])
    
    # Base case
    min1, color1, min2, color2 = self.twoMins(costs[0])
    
    # Recursive case
    for i in range(1, n):
        # 1-- initialize min1, color1, min2, color2 for the next round
        Color1, Color2 = 0, 1
        Min1 = costs[i][Color1] + (min1 if Color1 != color1 
                                        else min2 )
        Min2 = costs[i][Color2] + (min1 if Color2 != color1
                                        else min2 )
        
        if Min1 > Min2:
            Color1, Min1, Color2, Min2 = Color2, Min2, Color1, Min1
        
        # 2-- scan the costs for this row, and update them correspondingly
        for thisColor in range(2, k):
            thisCost = costs[i][thisColor] + (min1 if thisColor != color1
                                                    else min2)
            if thisCost < Min1:
                Min2, Color2 = Min1, Color1
                Min1, Color1 = thisCost, thisColor
            elif thisCost < Min2:
                Min2, Color2 = thisCost, thisColor
        
        # 3-- go to the next iteration
        min1, color1, min2, color2 = Min1, Color1, Min2, Color2
        
    return min1
````

````python
# return all the permutations of s including s where s can contain duplicate symbos
# Use a set 
def permute(self, s):
    # Base case
    if len(s) == 0: return set()
    if len(s) == 1: return set([s])

    # Recursive case
    fullPerms = set()
    subPerms = self.permute(s[1:])
    for subPerm in subPerms:
        for i in range(len(subPerm) + 1):
            fullPerm = subPerm[:i] + s[0] + subPerm[i:]
            fullPerms.add(fullPerm)
    return fullPerms


def generatePalindromes(self, s: str) -> List[str]:
    # handle degenerate case
    if len(s) == 1: return [s]
    freq = collections.Counter(s)
    odds = 0
    oddChar = None
    for c in freq:
        if freq[c] % 2 == 1: 
            odds += 1
            oddChar = c
    if odds > 1: return []
    
    # Build a sample left half containing all chars
    sample = ""
    for c in freq:
        sample += c * (freq[c] // 2)
    
    # Generate all possbile left halves
    allLefts = self.permute(sample)
    
    # Piece them together to form a complete palindrome
    allPerms = []
    for leftHalf in allLefts:
        if oddChar is not None:
            allPerms.append(leftHalf + oddChar + leftHalf[::-1])
        else:
            allPerms.append(leftHalf + leftHalf[::-1])
    return allPerms
````

### 270 Closest Binary Search Tree Value
 - go left if target is smaller than current root value, and go right otherwise. Keep track of the closest to target value at each step
 - $BinarySeaerch$
    - Search space `all the nodes in the tree`. At each step, eliminate roughly half by going either left or the right
    - If `target` greater than cur val, then right subtree eliminated
    - If `target` smaller than cur val, then left subtree eliminated

````python
def closestValue(self, root: Optional[TreeNode], target: float) -> int:
    # --- return the one closer one betwen {val1, val2} and target
    def getCloser(val1, val2):
        diff1 = abs(val1 - target)
        diff2 = abs(val2 - target)
        return val1 if diff1 < diff2 else val2
    
    closest = -math.inf
    while root:
        closest = getCloser(closest, root.val)
        if target == root.val: break
        elif target < root.val:
            root = root.left
        else:
            root = root.right
    return closest
````

### 276 Paint Fence
 - DP
 - Casing for painting `[i]` = the same color as `[i-1]` + a different color than `[i-1]`
 - Recurrence
 $$DP_i = DP_{i-1} * (k-1) + DP_{k-2} * (k-1)$$ 

````python
class Solution {
public:
    int numWays(int n, int k) {
        // Handle degenerate cases
        if (n == 0) return 0;
        if (n == 1) return k;
        if (n == 2) return k * k;
        
        vector<int> DP(n+1, 0); // DP[i] = the number of ways to paint the first i posts
        
        // Base case
        DP[0] = 0;
        DP[1] = k;
        DP[2] = k * k;
        
        // Recursive case
        for (int i = 3; i <= n; i++) {
            // case 1: i-1 and i-2 use the same color
            //      => have k-1 choices for each color taken by i-3
            DP[i] += (k-1) * DP[i-2];
            // case 2: i-1 and i-2 use different colors
            //      => have k-1 choices for each color taken by i-2
            DP[i] += (k-1) * DP[i-1];
        }
        
        // Return
        return DP[n];
    }
};
````

### 277 Find the Celebrity
 - Pattern: 
   - Query mode: each `query(x, y)` can eliminate either `x` or `y`
   - Want to query in a way to avoid no redundant elimination $\implies$ only need `n-1` queries

````python
def findCelebrity(self, n: int) -> int:
    candidate = 0
    
    # eliminate n-1 people
    for i in range(1, n):
        # invariant1: candidate < i at the beginning of each iteration
        # invariant2: possible range is {candidate} U {i, i+1, ..., n-1)
        # invariant3: candidate is the only one possible out of {0, 1, ..., i-1}
        if knows(candidate, i):
            candidate = i
            
    # varify the remaining one
    for i in range(n):
        if i == candidate: continue
        if knows(candidate, i):
            return -1
        if not knows(i, candidate):
            return -1
    return candidate
````

### 279 Perfect Squares

 - Iterating through all square numbers <= n

````python
#include <bits/stdc++.h>

class Solution {
public:
    void display(set<int> &S) {
        set<int>::iterator it;
        
        for (it = S.begin(); it != S.end(); it++) {
            cout << *it << " ";
        }
        cout << endl;
    }
    
    int numSquares(int n) {
        int i;
        vector<int> squares;
        
        // a vector of squares numbers <= n
        for (i = 1; i <= pow(n, 0.5); i++) {
            squares.push_back(i * i);
        }
        
        // Using BFS
        int level = 0;
        set<int> frontier;
        frontier.insert(n);
        
        while (frontier.size() > 0) {
            level++;
            set<int> next_frontier;
            for (set<int>::iterator it = frontier.begin(); it != frontier.end(); it++) {
                int remain = *it;
               
                for (auto square : squares) {
                    if (remain == square) return level;
                    else if (remain < square) break; // no need to traverse all squares
                    else {
                        next_frontier.insert(remain - square);
                    }
                }
            }
            frontier = next_frontier;
        }
        return level;
    }
};
````

### 280 Wiggle Sort
 - Justification: 
 - if `A <= B < C`, then swap `B` with `C` we have `A <= C >= B` must satify because
    - `A <= C` holds becasue `A <= B < C` implies `A <= C`
    - `C >= B` holds because `B < C` implies `C >= B`

 - if `A >= B > C`, then swap `B` with `C` we have `A >= C  B` must satify because
    - Similar reasnos as before
 - Conclusion: only needs to swap neighboring pairs

````python
def wiggleSort(self, nums: List[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    leq = lambda x, y : x <= y
    geq = lambda x, y : x >= y
    relationship = (leq, geq)
    
    for i in range(len(nums)-1):
        # 1-- choose the cmp for (i, i+1)
        cmp = relationship[i % 2]
        
        # 2-- swap if the cmp relationship is not satisfied
        if not cmp(nums[i], nums[i+1]):
            nums[i], nums[i+1] = nums[i+1], nums[i]
````

### 281 Zigzag Iterator
 - 思路: traverse all the vectors in a zig-zag manner -> $\text{Last In First Out}$
 - Keep track of a `Queue` containing entries in the form of `(vec_idx, ele_idx)`
 - Every time `popleft()` from the queue and `append` to the right is there are remaining elements in the current vec

````python
class ZigzagIterator:
    def __init__(self, v1: List[int], v2: List[int]):
        self.ptrs = deque([]) # containing (vec_idx, ele_idx)
        self.arrays = [v1, v2]
        if len(v1) > 0:
            self.ptrs.append((0, 0))
        if len(v2) > 0:
            self.ptrs.append((1, 0))
        
    def next(self) -> int:
        vec_idx, ele_idx = self.ptrs.popleft()
        ele = self.arrays[vec_idx][ele_idx]
        
        # move on to the next element of the current vector, if there is any
        if ele_idx+1 < len(self.arrays[vec_idx]):
            self.ptrs.append((vec_idx, ele_idx + 1))
        
        return ele
        
    def hasNext(self) -> bool:
        return len(self.ptrs) > 0
````

### 286 Walls and Gates
 - 思路 Perform BFS search where 
    - `from` the initial frontier is all the gates
    - `via` can only go through 
 - `反向search` ->Think about searching from the destination to the starting place
 - 表示empty room的INF正好可以用作unvisited room的距离 i.e.the distance is stored in `room[r][c]` itself

````python
def wallsAndGates(self, rooms: List[List[int]]) -> None:
        m = len(rooms)
        n = len(rooms[0])
        INF = 2147483647
        
        def withinBound(r, c):
            return 0 <= r < m and 0 <= c < n
        
        def BFS():
            # initialize the frontier to be all the gates
            Q = deque()
            for r in range(m): 
                for c in range(n):
                    if rooms[r][c] == 0:
                        Q.append((r, c))
            
            # Perform BFS 
            while Q:
                # visit the current node
                r, c = Q.popleft()
                
                # add neighbors to the next frontier
                for (dR, dC) in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    R = r + dR
                    C = c + dC
                    if withinBound(R, C) and rooms[R][C] == INF:
                        rooms[R][C] = min(rooms[R][C], 1 + rooms[r][c])
                        Q.append((R, C))
        
        # driver    
        BFS()
````

### 289 Game of Life
 - Store the information about the new state and the old state at the same time using no more bits
 - New encoding system in the comment

````python
class Solution:
def gameOfLife(self, board: List[List[int]]) -> None:
    """
    Do not return anything, modify board in-place instead.
    """
    rows = len(board)
    cols = len(board[0])

    # new encoding of the state that is compatible with the old 
    # encoding system
    # 'OLD'
    # die -> ?       0
    # live -> ?      1
    
    # 'NEW'
    # live -> live   1
    # live -> die    -1
    # die -> live    2
    # die -> die     0
    
    # criteria
    def used_to_live(r, c): return abs(board[r][c]) == 1 
    def now_live(r, c): return board[r][c] > 0
    def within(r, c): return 0 <= r < rows and 0 <= c < cols
    
    for r in range(rows):
        for c in range(cols):
            liveNbors = 0
            # count the neighboring cells of this (r, c) cell
            for dRow in [-1, 0, 1]:
                for dCol in [-1, 0, 1]:
                    if dRow == 0 and dCol == 0: continue
                    newR = r + dRow
                    newC = c + dCol
                    if within(newR, newC) and used_to_live(newR, newC):
                        liveNbors += 1
            # assign new state 
            if board[r][c] == 1:
                if (liveNbors < 2 or liveNbors > 3):
                    board[r][c] = -1 # live -> die
                else:
                    board[r][c] = 1 # live -> live
            else:
                if liveNbors == 3:
                    board[r][c] = 2 # die -> live
                else:
                    board[r][c] = 0 # die -> die
    
    # restore to the old encoding system
    for r in range(rows):
        for c in range(cols):
            if now_live(r, c):
                board[r][c] = 1
            else:
                board[r][c] = 0
````

### 290 Word Pattern
 - For handling bijections -> keep track of two mappings `f` and `g`
 - Three cases
    - Case 1: `w` in `f` and `s` in `g` ->  check `f[w] == s and g[s] == w`
    - Case 2: `w` not in `f` and `s` not in `g` -> add `f[w] = s and g[s] = w`
    - Case 3: one in, one not in -> return False

````python
def wordPattern(self, pattern: str, s: str) -> bool:
    words = s.split()
    symbols = list(pattern)
    f = dict() # from word to symbol
    g = dict() # from symbol to word
    
    if len(words) != len(symbols):
        return False
    
    for w, s in zip(words, symbols):
        if w in f and s in g: 
            if f[w] != s or g[s] != w:
                return False
        elif w not in f and s not in g:
            f[w] = s
            g[s] = w
        else: return False

    return True
````

### 291 Word Pattern II
 - Backtrack
 - 对 `cursor` 概念的理解 : 
    - `s_start` 指可以从 `s[s_start : ?]` 开始构建下一个 `w` 尝试进行map
    - `p_cursor`指当前的`s[s_start : ?]` 对应的pattern是`ps[p_cursor]`
 - 由于 `s_end` 才用exclusive方式 -> 下一个 `backtrack`的argument为`backtrack(s_end, p_cursor+1)`

````python
def wordPatternMatch(self, pattern: str, s: str) -> bool:
    ps = list(pattern)
    
    # try matching from s_cursor and p_cursor
    def backtrack(s_start, p_cursor, w2p, p2w):
        # Base case
        if s_start == len(s) and p_cursor == len(ps): 
            return True
        if s_start >= len(s) or p_cursor >= len(ps):
            return False
        
        # Recursive case
        for s_end in range(s_start+1, len(s)+1):
            w = s[s_start : s_end]
            p = ps[p_cursor]
            # -- Case 1
            if w in w2p and p in p2w:
                if p == w2p[w] and w == p2w[p]: 
                    if backtrack(s_end, p_cursor+1, w2p, p2w): return True
            # -- Case 2
            elif w not in w2p and p not in p2w:
                w2p[w] = p
                p2w[p] = w
                if backtrack(s_end, p_cursor+1, w2p, p2w): return True
                del w2p[w]
                del p2w[p]
        return False
    
    # driver
    w2p = dict()
    p2w = dict()
    res = backtrack(0, 0, w2p, p2w)
    return res
````

### 294 Flip Game II
 - Dynamic Programing with Memoization
 - Problem definition `DP[s] = whether starting player can win with s`
 - Recursive case: Try all the replacement results from `s` -> `newS` that we can change to. Make sure `DP(newS) = False` -> the opponent cannot win -> we win

````python
def canWin(self, currentState: str) -> bool:
    cache = dict()
    
    # returns if the starting player can guarantee to win with s
    def DP(s):
        # Base case 1
        if s in cache: return cache[s]
        if '++' not in s: 
            cache[s] = False
            return False
        # Recursive case
        for i in range(len(s)-1):
            if s[i] == '+' and s[i+1] == '+':
                newS = s[:i] + '--' + s[i+2:]
                # make the change so s becomes newS
                # want to make sure opponent CANNOT win
                if not DP(newS):
                    cache[s] = True
                    return True
        cache[s] = False
        return False
    
    # driver
    res = DP(currentState)
    return res
````

 - First pass: greedily harvest all the bulls, and record all the unmatched chars in the actual string into the dictionary `unmatched[c] -> c's unmatched number of instances`
 - Second pass: go over the string again, for each unmatched char in guess `c`, if `c` is in `unmatched` -> we find a cow `cow += 1` and consume one instance `unmatched[c] -= 1`

````python
def getHint(self, secret: str, guess: str) -> str:
    bulls = 0
    cows = 0
    unmatched = dict()
    
    # compute bulls and calculate `unmatched`
    for c1, c2 in zip(secret, guess):
        if c1 == c2:
            bulls += 1
        else:
            unmatched[c2] = unmatched.get(c2, 0) + 1
            
    # compute cows
    for c1, c2 in zip(secret, guess):
        if c1 != c2:
            if unmatched.get(c1, 0) > 0:
                cows += 1
                unmatched[c1] -= 1
    
    # final
    return str(bulls) + 'A' + str(cows) + 'B'
````

### 300 Longest Increasing Subsequence
- Define `DP[i]` := max length of LIS that ends at index i
- Recurrence $DP[i] = \max_{j < i: arr[j] < arr[i]} 1+DP[j]$

````python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Define DP[i] := max length of LIS that ends at index i
        n = len(nums)
        DP = [1] * n
        
        # Base Case
        DP[0] = 1
        
        # Recursive Case
        for i in range(1, n):
            curr = nums[i]
            for j in range(i):
                if nums[j] < curr:
                    DP[i] = max(DP[i], 1+DP[j])

        # take the max of all ending indices
        ans = 0
        for i in range(n):
            ans = max(ans, DP[i])
        return ans
````
