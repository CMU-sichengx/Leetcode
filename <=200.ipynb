- Note: root (pointer address) can be used as entries for a deque

````python
import queue
from collections import deque
class Solution(object):
    def BFS(self, frontier, acc):
        # Base Case
        if len(frontier) == 0:
            return
        
        # Recursive Case
        # record current frontier
        vals = []
        for node in frontier:
            vals.append(node.val)
        acc.append(vals)
        # construct new frontier
        newFrontier = []
        for node in frontier:
            if node.left is not None:
                newFrontier.append(node.left)
            if node.right is not None:
                newFrontier.append(node.right)
         # recurse
        self.BFS(newFrontier, acc)
        
    def BFS_Queue(self, root):
        levels = []
        Q = deque()
        Q.append(root) 
        
        while Q:
            # record the current level
            # build the next frontier at the same time
            level = []
            level_n = len(Q)
            for _ in range(level_n):
                node = Q.popleft()
                level.append(node.val)
                if node.left: Q.append(node.left)
                if node.right: Q.append(node.right)
            # accumulate all the levels
            levels.append(level)
        
        return levels


````

### 104 Maximum Depth of Binary Tree

````python
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # Base Case
        if root is None:
            return 0
        # Recursive Case
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
````

### 105 Construct Binary Tree from Preorder and Inorder Traversal
- `Preorder` always starts with the root node $\implies$ allows us to separate
the linear array into segments and call them recursively

````python
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """
        # The first element of preorder is always the correct root of the tree
        # Base Case:
        if len(preorder) == 0:
            return None
        
        # Recursive Case
        rootVal = preorder[0]
        cut = inorder.index(rootVal)
        lSize = cut
        # segments for recursive arguments
        preorderLeft = preorder[1 : 1+lSize]
        preorderRight = preorder[1+lSize : ]
        inorderLeft = inorder[0 : cut]
        inorderRight = inorder[cut+1 : ]
        # recurse
        tree = TreeNode()
        tree.val = rootVal
        tree.left = self.buildTree(preorderLeft, inorderLeft)
        tree.right = self.buildTree(preorderRight, inorderRight)
        return tree
````

### 109 Convert Sorted List to Binary Search Tree

````python
class Solution:
    # This function cuts `head` in half and return the second half
    # the first half is still represented by `head`
    def split_half(self, head):
        prevPtr = None
        slowPtr = head
        fastPtr = head
    
        while fastPtr and fastPtr.next:
            prevPtr = slowPtr
            slowPtr = slowPtr.next
            fastPtr = fastPtr.next.next
       
        if prevPtr:
            prevPtr.next = None

        return slowPtr
        
    def sortedListToBST(self, head: Optional[ListNode]) -> Optional[TreeNode]:
        # Handle base case
        if head is None:
            return None
        if head.next is None:
            return TreeNode(head.val)
        
        # Handle recursive case
        mid = self.split_half(head)
        root = TreeNode(mid.val)
        root.left = self.sortedListToBST(head)
        root.right = self.sortedListToBST(mid.next)
        return root
````

### 111 Minimum Depth of Binary Tree

````python
class Solution:
    def minDepth(self, root: Optional[TreeNode]) -> int:
        # base case
        if root is None:
            return 0
        
        # recursive case
        if root.left is None and root.right is None:
            return 1
        if root.left is None:
            return self.minDepth(root.right) + 1
        if root.right is None:
            return self.minDepth(root.left) + 1
        else:
            min_l = self.minDepth(root.left)
            min_r = self.minDepth(root.right)
            return 1 + min(min_l, min_r)
````

### 115 Distinct Subsequences
 - Simple dynamic programming
 - Sub problem `DP[i, j] = the numebr of occurences of t[:i] in s[:j]`
 - Recursive formula:
    - if currently `t[i] == s[j]`
        - case 1: use `t[i]` to match `s[j]` => `DP[i-1][j-1]`
        - case 2: do not leverage `s[j]` => `DP[i-1][j-1]`
    - if currently `t[i] != s[j]`
        - case 1: do not leverage `s[j]` => `DP[i-1][j-1]`

````python
def createMatrix(self, rowN, colN, init):
    return [ [init for _ in range(colN)] for _ in range(rowN) ]

def numDistinct(self, s: str, t: str) -> int:
    # DP[i, j] = the number of distinct substrings of t[:i] in s[:j]
    # i: the cursor in t
    # j: the cursor in s
    DP = self.createMatrix(len(t)+1, len(s)+1, 0)
    
    # Base case
    DP[0][0] = 1 # should not be using this line
    
    # filling the first column
    for i in range(1, len(t)): DP[i][0] = 0
    
    # filling the first row
    for j in range(1, len(s)): DP[0][j] = 0
    
    DP[1][1] = 1 if s[0] == t[0] else 0
    
    for j in range(2, len(s)+1):
        DP[1][j] = DP[1][j-1] + (1 if s[j-1] == t[0] else 0)
    
    print(DP)
    
    # Recursive case
    for i in range(2, len(t)+1):
        for j in range(1, len(s)+1):
            DP[i][j] = 0
            if t[i-1] == s[j-1]:
                # case 1: use t[i] to match s[j]
                DP[i][j] = DP[i-1][j-1] + DP[i][j-1]
            else:
                DP[i][j] = DP[i][j-1]

    return DP[len(t)][len(s)]
````

### 117 Populating Next Right Pointers in Each Node II
 - Finish level by level 
 - Establish links in level `n+1` by traversing level `n`
 - Difference from 116: no longer a perfect binary search tree
    - keep track of `nextLeftmost`
    - keep track of `below` and use while loop to find `below`'s next
    ```

                    above = above.next
    ```

````python
def connect(self, root: 'Optional[Node]') -> 'Optional[Node]':
    # Handle degenerate case
    if root is None:
        return root
    
    leftmost = root
    while leftmost is not None:
        above = leftmost
        nextLeftmost = None
        while above:
            # --- Update nextLeftmost
            if above.left and not nextLeftmost: nextLeftmost = above.left
            if above.right and not nextLeftmost: nextLeftmost = above.right
            # --- establish connection 1
            if above.left and above.right:
                above.left.next = above.right
            below = above.right if above.right else above.left
            # --- (A) establish connection 2
            if below: 
                above = above.next
                while above and (not above.left and not above.right):
                    above = above.next
                if above: below.next = above.left if above.left else above.right
                else: below.next = None
            # --- (B) go to the next place
            else: above = above.next

        leftmost = nextLeftmost
    return root
````

````python
def getRow(self, rowIndex: int) -> List[int]:
    # 1-- handle degenerate cases
    if rowIndex == 0: return [1]
    
    # 2-- handle > 0 indexed rows
    row = [1, 1]
    for _ in range(rowIndex-1):
        newRow = [1]
        for i in range(len(row)-1):
            newRow.append(row[i] + row[i+1])
        newRow.append(1)
        row = newRow
    
    # 3-- finish and return
    return row
````

````python
def minimumTotal(self, triangle: List[List[int]]) -> int:
    rowN = len(triangle)
    colN = len(triangle[-1])
    
    DP = self.createMatrix(rowN, colN, 0)
    DP = triangle
    # -- DO[i][j] = the min path sum from top to [i][j]
    
    # Base case
    DP[0][0] = triangle[0][0]
    for r in range(1, rowN):
        DP[r][0] = DP[r-1][0] + triangle[r][0]
        DP[r][r] = DP[r-1][r-1] + triangle[r][r]
    
    # Recursive case
    for r in range(2, rowN):
        for c in range(1, r):
            DP[r][c] = triangle[r][c] + min(DP[r-1][c], DP[r-1][c-1])
    
    return min(DP[rowN-1])
````

### 121 Best Time to Buy and Sell Stock
- _time-series array_ 思路: $$\textit{an array = a sequences of valleys and peaks}$$
- Calculate `min_price_so_far` for each day $Scan$
- For each day `i`, assuming we will sell that day, `profit = price[i] - min_so_far[i]`
- Iterate through `profit[]` to get the max

````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)
        
        # calculate minSoFar for each day
        minByDay = [0] * n
        minByDay[0] = prices[0]
        for i in range(1, n):
            minByDay[i] = min(minByDay[i-1], prices[i])
        
        # calculate profit for each day
        profitByDay = [0] * n
        for i in range(n):
            profitByDay[i] = prices[i] - minByDay[i]
            
        # extract the max
        res = 0
        for i in range(n):
            res = max(res, profitByDay[i])

        return res
````

### 123 Best Time to Buy and Sell Stock III

````python
def maxProfit(self, prices: List[int]) -> int:
    n = len(prices)
    
    # -- leftProfits[i] = max one-transaction profit considering only [:i+1]
    # i ranges 0, 1, ..., n-1
    profitsUntil = [0] * n
    profitsUntil[0] = 0
    minBuySoFar = prices[0]
    
    for d in range(1, n):
        profitsUntil[d] = max(profitsUntil[d-1], prices[d] - minBuySoFar)
        minSoFar = min(minBuySoFar, prices[d])
    
    # -- profitsFrom[i] = max one-transaction profit considering only [i:]
    # i ranges 0, 1, ..., n-1
    profitsFrom = [0] * (n)
    profitsFrom[n-1] = 0
    maxSellSoFar = prices[n-1]
    
    for d in range(n-2, -1, -1):
        profitsFrom[d] = max(profitsFrom[d+1], maxSellSoFar - prices[d])
        maxSellSoFar = max(maxSellSoFar, prices[d])
    
    # -- Get max among all division points
    maxProfit = 0
    for mid in range(n):
        thisProfit = profitsUntil[mid] + profitsFrom[mid]
        maxProfit = max(maxProfit, thisProfit)
    return maxProfit
````

### 124 Binary Tree Maximum Path Sum
 - Function model: used when the desired tree object does not necessarily bound to the overall root of the tree
 - Write a function `getResultByRoot` and a global variable `bestOverAllRoots` that 
    - returns the final result `under the constraint that it can be recursively used by its parent callers`
    - modifies the `bestOverAllRoots` without the constraints above

````python
def maxPathSum(self, root: Optional[TreeNode]) -> int:
    globalMax = [-math.inf]
    '''
    (1) returns the max path sum within the subtree rooted at root
    (2) modify the global max to record the max value
    '''
    def maxPathFromNode(root):
        # Base case
        if root.left is None and root.right is None: 
            globalMax[0] = max(globalMax[0], root.val)
            return root.val

        # Recursive case
        leftMax = maxPathFromNode(root.left) if root.left else -math.inf
        rightMax = maxPathFromNode(root.right) if root.right else -math.inf
        
        # For purpose (1)
        globalMax[0] = max(
            globalMax[0],
            root.val,
            leftMax, 
            rightMax, 
            leftMax + root.val, 
            rightMax + root.val,
            leftMax + rightMax + root.val
        )
        
        # For purpose (2)
        return max(
            root.val,
            leftMax + root.val, 
            rightMax + root.val,
        )
    
    maxPathFromNode(root)
    return globalMax[0]
````

### 125 Valid Palindrome
- Two pointer method
- Always move the pointers `l` and `r` until they point to alpha-numeric chars
- Remeber `l < n` and `r >= 0` when moving pointers

````python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """
        n = len(s)
        s = s.lower()
        l = 0
        r = n - 1
        # move l and r to the first alpha numeric 
        while l < n and not s[l].isalnum():
            l += 1
        while r >= 0 and not s[r].isalnum():
            r -= 1
        while (l <= r):
            if s[l] != s[r]:
                return False
            # move l and r until they each point to an alphanumeric
            l += 1
            r -= 1
            while l < n and not s[l].isalnum():
                l += 1
            while r >= 0 and not s[r].isalnum():
                r -= 1
        return True
````

### 128 Longest Consecutive Sequence
- Use the set data structure when we only care about `existence of an element` and don't care about the order or count.
- Reduce each consecutive sequence to its `Head` element $\implies$ `num-1 not in set`
- For each starting element `start`, repeatedly ask if `start+1` is in `set` or not, record `curlen`
- Keep the `maxlen` by updating it against `curlen`

````python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        s = set(nums)
        maxLen = 0
        for num in nums:
            if num-1 not in s: # if num is the start of a sequence
                curLen = 1
                target = num + 1
                while target in s:
                    curLen += 1
                    target += 1
                maxLen = max(maxLen, curLen)
        return maxLen
                
````

### 134 Gas Station
 - Circular MCSS to get the start index
  - empty subsequence [] not allowed 
  - update start and end along the way
  
 - Try from the best starting index and simulate the tour

````python
import math
class Solution:
    def MaxCSS(self, nums):
        DP = [0] * len(nums)
        
        # find MaxCSS of the `netChange` array 
        DP[0] = nums[0]
        start, end = 0, 0 # [start, end] both inclusive
        
        for i in range(1, len(nums)):
            DP[i] = max(
                nums[i],
                nums[i] + DP[i-1]
            )
            
            if DP[i] == nums[i]:
                start, end = i, i
            else:
                end = i
        return (start, end)
            
        
    def MinCSS(self, nums):
        nums = list(map(lambda x:-x, nums))
        start, end = self.MaxCSS(nums)
        return start, end
    
    # return the start of the MCS for the `nums` array where wrap-around is allowed
    def cirMCSS(self, nums):
        start1, end1 = self.MaxCSS(nums)
        withoutWrapAround = sum(nums[start1:end1+1])
        start2, end2 = self.MinCSS(nums)
        withWrapAround = sum(nums) - sum(nums[start1:end1+1])
        
        if withoutWrapAround >= withWrapAround:
            return start1
        else:
            return (end2 + 1) % len(nums)
        
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        net_change = [0] * len(gas)
        for i in range(len(gas)):
            net_change[i] = gas[i] - cost[i]
            
        best_start_idx = self.cirMCSS(net_change)
        i = best_start_idx
        tank = 0
        
        for _ in range(len(net_change)):
            # print(i)
            tank += net_change[i]
            if tank < 0:
                return -1
            i = (i + 1) % len(net_change)

        return best_start_idx
````

````python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        DP[l]: if the l-prefix of the word s can be combined using wordDict
        """
        n = len(s)
        corpus = set(wordDict) # speed up the operation
        DP = [False] * (n+1)
        
        # Base Case
        DP[0] = True
        DP[1] = s[0] in corpus
        
        # Recursive Case
        for l in range(2, n+1):
            for lastWordLen in range(1, l+1):
                lastWord = s[l-lastWordLen : l]
                DP[l] = DP[l] or (lastWord in corpus and DP[l-lastWordLen])
        return DP[n]
````

### 144 Binary Tree Preorder Traversal

````python
def display(self, node):
    if node is None: print('None')
    else: print(node.val)

def showList(self, L):
    print('-----')
    for node in reversed(L): self.display(node)
    print('-----')

def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
    paramStack = [root]
    output = []
    
    while paramStack:
        node = paramStack.pop()
        if node is None: 
            continue
        else:
            output.append(node.val)
            # last recursive call
            paramStack.append(node.right)
            # first recursive call
            paramStack.append(node.left)
        
    return output
````

### 147 Insertion Sort List
 - For linked lists, $O(1)$ space 不一定 in-place.
    - 可以先构建一个或者多个dummy head node，不断将原link中的nodes install到dummy上
 - Keep track of `to_insert` as a cursor pointing to the node `in the original linked list`
 - For each `to_insert`, let `prev_to_insert` traverse the dummy node and once found a place or reach the end `of the dummy link`,
    - delete `to_insert` fro the `original link` and install it in the `dummy link`

````python
def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    # 1-- handle degenerate case
    dummy = ListNode(-math.inf)
    
    # 2-- sort the array
    to_insert = head
    while to_insert:
        # record next
        to_insert_next = to_insert.next
        
        # locate the place to insert
        prev_to_insert = dummy
        while prev_to_insert.next and prev_to_insert.next.val < to_insert.val:
            prev_to_insert = prev_to_insert.next
        
        # install the node       
        to_insert.next = prev_to_insert.next
        prev_to_insert.next = to_insert
        
        # move on to the next yet-to-sorted node
        to_insert = to_insert_next
    
    return dummy.next
````

### 152 Maximum Product Subarray
- DP recurrence: 
- $maxByIndex[i] = max(val, val*maxByIndex[i-1], val*minByIndex[i-1])$
- $minByIndex[i] = min(val, val*maxByIndex[i-1], val*minByIndex[i-1])$


````python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        n = len(nums)
        maxByIndex = [0] * n
        minByIndex = [0] * n
        
        # Initialize base case
        maxByIndex[0] = nums[0]
        minByIndex[0] = nums[0]
        bestSoFar = nums[0]
        
        # Traverse the array
        for i in range(1, n):
            val = nums[i]
            maxByIndex[i] = max(val, val*maxByIndex[i-1], val*minByIndex[i-1])
            minByIndex[i] = min(val, val*maxByIndex[i-1], val*minByIndex[i-1])
            bestSoFar = max(bestSoFar, maxByIndex[i])
        
        return bestSoFar
````

### 153 Find Minimum in Rotated Sorted Array
- Binary search
```
lo, hi = 0, n
while lo < hi:
    mid = lo + (hi-lo) // 2
    ...
```
1 Find the rotation index `i` $\implies$
2 The `min` element is situated at the `i+1` 

````python
class Solution:
    # find the rotation index
    def binsearch_rotIndex(self, nums):
        n = len(nums)
        lo = 0
        hi = n
        while lo < hi:
            # invariant: rotIndex potentially within range [lo, hi)
            mid = lo + (hi-lo) // 2
            if mid == n-1:
                if nums[mid-1] > nums[mid]:
                    return mid-1
                else:
                    return -1
            if nums[mid] > nums[mid+1]:
                return mid
            else:
                if nums[mid] < nums[0]:
                    hi = mid
                else:
                    lo = mid + 1
        return -1 # will not reach this line
        
    def binsearch(self, nums, target):
        n = len(nums)
        lo = 0
        hi = n
        while (lo < hi):
            mid = lo + (hi-lo) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                lo = mid + 1
            else:
                hi = mid
        return -1 # will not reach this line
        
    
    def findMin(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        rotIndex = self.binsearch_rotIndex(nums)
        return nums[rotIndex + 1]
````

### 154 Find Minimum in Rotated Sorted Array II
 - Binary search (Two triangle模型变种)
 - Casing: compare `nums[mid]` vs `nums[hi]`.
    - 不比较`nums[0]`的原因 -> 相等的时候可以将 `hi -= 1` 保证缩小search space
    - 当`nums[mid] == nums[hi]` -> `hi -= 1`

````python
def findMin(self, nums: List[int]) -> int:
    n = len(nums)
    lo, hi = 0, n-1
    
    # invariant: pivot in [lo, hi]
    while lo < hi:
        mid = (lo + hi) // 2
        midVal = nums[mid]
        hiVal = nums[hi]
        # 1-- if midVal > hiVal, then mid on first triangle, min on the second triangle
        if midVal > hiVal: 
            lo = mid + 1
        # 2-- if midVal < hiVal, then mid and min both on the second triangle
        elif midVal < hiVal:
            hi = mid
        # 3-- if midVal == hiVal, not sure which triangle mid is in
        #  -- but is sure that hi itself is not the min
        else:
            hi = hi - 1
        
    # at this point lo = hi converges to one point
    return nums[lo]
````

### 156 Binary Tree Upside Down
 - Edge Case: Initially set the `root.left = root.right = None`.
 - Pattern: Keep track of `X, Y, Z` and update with `newX ,newY, newZ` to avoid reasoning about different pointers

````python
def upsideDownBinaryTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
    if root is None: return None
    
    X, Y, Z = root, root.left, root.right
    X.left = None
    X.right = None
    
    while Y:
        # -- keep for record
        newX = Y
        newY = Y.left
        newZ = Y.right
        # -- perform reconnecting
        Y.left = Z
        Y.right = X
        # -- to the next iteration
        X, Y, Z = newX, newY, newZ
    return X
````

### 159 Longest Substring with At Most Two Distinct Characters
 - Warning: checking `num in freq` and checking `freq[num] changes from 0 to 1` is different ! Use the later.
 - Sliding window with `freq` and `cnt`. 

````python
def lengthOfLongestSubstringTwoDistinct(self, s: str) -> int:
    begin, end, distincts, freq = 0, 0, 0, dict()
    length = 0
    
    while end < len(s):
        # ---------- 1
        # move 'end' to right by 1 and update correspondingly
        endChar = s[end]
        freq[endChar] = freq.get(endChar, 0) + 1
        if freq[endChar] == 1: distincts += 1
        end += 1
        
        # ---------- 2
        # keep moving 'begin' to right until the window is restored again
        while distincts > 2:
            beginChar = s[begin]
            freq[beginChar] -= 1
            if freq[beginChar] == 0:
                distincts -= 1
            begin += 1
            
        # ---------- 3
        # now the window is restored, update answer with wrt this window
        # print(begin, end, s[begin:end])
        length = max(length, end-begin)
        
    return length
````

### 161 One Edit Distance
 - 审题 : needs to be exactly `one` edit distance
 - Differentiate the shorter one and the longer one
 ```
        # adjust so that s is always the shorter one
        if len(s) > len(t): s, t = t, s
 ```

````python
def isOneEditDistance(self, s: str, t: str) -> bool:
    if abs(len(s) - len(t)) > 1: return False
    
    # 1-- Case: same length 
    if len(s) == len(t):
        diff = 0
        for c1, c2 in zip(s, t):
            if c1 != c2: diff += 1
        if diff != 1: return False
        else: return True
    # 2-- Case: length differs by one
    else:
        # adjust so that s is always the shorter one
        if len(s) > len(t): s, t = t, s
            
        # detect the first different index
        i = 0
        while i < len(s) and i < len(t) and s[i] == t[i]: i += 1
            
        # check for equality
        return s[i:] == t[i+1:]
````

### 162 Find Peak Element

````python
def findPeakElement(self, nums: List[int]) -> int:
    n = len(nums)
    
    lo, hi = 0, n
    while lo < hi:
        mid = (lo + hi) // 2
        midNum = nums[mid]
        leftNum = -math.inf if mid == 0 else nums[mid-1]
        rightNum = -math.inf if mid == n-1 else nums[mid+1]
        
        if leftNum < midNum and midNum > rightNum: return mid
        elif leftNum > midNum: hi = mid
        elif midNum < rightNum: lo = mid + 1
    
    raise Exception("cannot be here")
````

### 165 Compare Version Numbers

    - `L = list(map(lambda, l))`
    - `L = list(filter(lambda, l))`

````python
def compareVersion(self, version1: str, version2: str) -> int:
    ver1 = list(map(lambda s: int(s), version1.split('.')))
    ver2 = list(map(lambda s: int(s), version2.split('.')))
    
    # adjust so that ver1 is always the shorter one
    i = 0
    while i < min(len(ver1), len(ver2)):
        v1 = ver1[i]
        v2 = ver2[i]
        if v1 < v2: return -1
        elif v1 > v2: return 1
        i += 1
        
    while i < len(ver1):
        if ver1[i] != 0: return 1
        i += 1
    
    while i < len(ver2):
        if ver2[i] != 0: return -1
        i += 1

    return 0
````

### 167 Two Sum II - Input Array Is Sorted
 - Keep two pointers one from beginning, one from end
 - Keep track of `A[l] + A[r]` and move `l` and `r` towards each other
 - Linear time
 

````python
def twoSum(self, numbers, target):
    """
    :type numbers: List[int]
    :type target: int
    :rtype: List[int]
    The two point method
    Matrix visualization
    """
    left = 0
    right = len(numbers)-1
    while (left < right):
        if numbers[left] + numbers[right] == target:
            return [left+1, right+1]
        elif numbers[left] + numbers[right] < target:
            left += 1
        else:
            right -= 1
    return [left+1, right+1]
````

### 168 Excel Sheet Column Title

````python
def convertToTitle(self, columnNumber: int) -> str:
    D = dict()
    for num in range(1, 27):
        D[num] = chr(num - 1 + ord('A'))

    columnName = ''
    while columnNumber > 0:
        remainder = columnNumber % 26
        if remainder == 0:
            columnName = 'Z' + columnName
            columnNumber = (columnNumber - 26) //  26
        else:
            columnName = D[remainder] + columnName
            columnNumber = (columnNumber - remainder) //  26
        
    return columnName
````

### 170 Two Sum III - Data structure design

### 173 Binary Search Tree Iterator
 - Inorder BST traversal using a stack
 - See `<=100.ipynb` for pseudocode
 - Invariant: the stack top always points to the currently explored smallest element so far
    - 1. pop the top node from the stack
    - 2. if the top node has a right child, go as left as possible along the right child
    - 3. return

````python
class BSTIterator:
    def __init__(self, root: Optional[TreeNode]):
        # -- invariant: the top element is the unvisited min element
        self.stack = []
        self.goLeftMost(root)
        
    def goLeftMost(self, node):
        while node:
            self.stack.append(node)
            node = node.left

    def next(self) -> int:
        # 1-- pop the topmost element from stack
        topNode = self.stack.pop()
        
        # 2-- by invariant, node.left has been processed or node.left is None
        # --- now process the node's right subtree
        if topNode.right: self.goLeftMost(topNode.right)
            
        # 3-- return the result
        return topNode.val
    
    def hasNext(self) -> bool:
        return len(self.stack) > 0
````

### 187 Repeated DNA Sequences

````python
def findRepeatedDnaSequences(self, s: str) -> List[str]:
    # handle degenerate cases
    if len(s) < 10: return []
    
    freq = dict() # mapping sequence to its freq
    repeats = []
    
    # initialize the window [lo, hi)
    lo, hi = 0, 10
    
    while hi <= len(s):
        # 1-- capture the current window
        seq = s[lo:hi]
        if seq in freq:
            if freq[seq] == 1: repeats.append(s[lo:hi])
            freq[seq] += 1
        else:
            freq[seq] = 1
        # 2-- move on to the next window
        lo += 1
        hi += 1
    return repeats
````

    - DP[l][1] = maximum value achievable for nums[0..l-1] when using nums[l-1]
    - DP[l][0] = maximum value achievable for nums[0..l-1] not using nums[l-1]
- Recurrence
    - $DP[i][0] = max(DP[i-1][0], DP[i-1][1])$
    - $DP[i][1] = nums[i-1] + DP[i-1][0]$
- Keep track of the last decision ！

````python
class Solution(object):
    def makeMatrix(self, rowN, colN, init):
        matrix = []
        for row in range(rowN):
            matrix.append([init] * colN)
        return matrix
            
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        Define DP[l][B] where B = 0 or 1
            DP[l][1] = maximum value achievable for nums[0..l-1] when using nums[l-1]
            DP[l][0] = maximum value achievable for nums[0..l-1] not using nums[l-1]
        """
        n = len(nums)
        DP = self.makeMatrix(n+1, 2, 0)
        
        # Handling edge cases
        if n == 1:
            return nums[0]
        if n == 2:
            return max(nums[0], nums[1])
        
        # Base Case
        DP[1][0] = 0
        DP[1][1] = nums[0]
        
        # Recursive Case
        for i in range(2, n+1):
            DP[i][0] = max(DP[i-1][0], DP[i-1][1])
            DP[i][1] = nums[i-1] + DP[i-1][0]
        
        return max(DP[n][0], DP[n][1])
````

### 199 Binary Tree Right Side View
 - Level Order Traversal using Queue

````python
def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
    # handle degenerate case
    if root is None: return []
    
    rightView = []
    Q = deque([root])
    while Q:
        rightView.append(Q[-1].val)
        n = len(Q)
        for _ in range(n):
            node = Q.popleft()
            if node.left: Q.append(node.left)
            if node.right: Q.append(node.right)
    return rightView
````

### 200 Number of Islands
- Backtracking
- Destructively modify the graph by setting `grid[row][col] = '!'`
- Loop through all pos `(r, c)`, if `grid[r][c] not conteminated`, search from `[r][c]` and increment counter

````python
class Solution(object):
    def BFS(self, grid, row, col, cnt):
        # Base Case
        rowN = len(grid)
        colN = len(grid[0])
        # out of bound
        if row < 0 or row >= rowN or col < 0 or col >= colN:
            return
        # not an island, reach the sea, stop
        if grid[row][col] != '1':
            return
        
        # Recursive Case
        # apply change and no need to restore it !
        grid[row][col] = '!' # this cell has been visited
        for (dRow, dCol) in [(-1,0), (1,0), (0,-1), (0,1)]:
            newRow = row + dRow
            newCol = col + dCol
            self.BFS(grid, newRow, newCol, cnt)
        
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        rowN = len(grid)
        colN = len(grid[0])  
        cnt = 0
        for row in range(rowN):
            for col in range(colN):
                if grid[row][col] == '1':
                    self.BFS(grid, row, col, cnt)
                    cnt += 1
        return cnt
                    
````
