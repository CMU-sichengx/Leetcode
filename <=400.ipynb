````python
class Solution {
public:
    int top, buttom, left, right;
    
    void dfs(vector<vector<char>>& image, int x, int y){
        // Base case
        if (x < 0 || x >= image.size() || y < 0 || y >= image[0].size() || image[x][y] == '0') return;
        
        // Recursive case
        image[x][y] = '0';
        top = min(top, x);
        buttom = max(buttom, x);
        left = min(left, y);
        right = max(right, y);
        
        dfs(image, x, y-1);
        dfs(image, x, y+1);
        dfs(image, x-1, y);
        dfs(image, x+1, y);
    }
    int minArea(vector<vector<char>>& image, int x, int y) {
        left = y;
        right = y;
        top = x;
        buttom = x;
        dfs(image, x, y);
        cout << left << " " << right << " " << top << " " << buttom << endl;
        return (right + 1 - left) * (buttom + 1 - top);
    }
};
````

### 303 Range Sum Query - Immutable
 - Prefix sum method
 - A *contiguous reduce sum of an array* -> *prefix sum*


````python
class NumArray:

def __init__(self, nums: List[int]):
    n = len(nums)
    prefSum = [0] * (n+1) # prefSum[i] = the sum of the first i numbers of nums
    
    for i in range(1, n+1):
        prefSum[i] = prefSum[i-1] + nums[i-1]
    
    self.prefSum = prefSum
    

def sumRange(self, left: int, right: int) -> int:
    return self.prefSum[right+1] - self.prefSum[left]
````

### 304 Range Sum Query 2D - Immutable
 - The 2D version of the prfix sum
 - `DP[rowN][colN]` = the sum of the first rowN rows and the first colN cols
 - Recurrence $DP[rowN][colN] = DP[rowN-1][colN] + DP[rowN][colN-1] + matrix[rowN-1][colN-1]$

````python
class NumMatrix {
public:
    vector<vector<int>> DP; 
    NumMatrix(vector<vector<int>>& matrix) {
        int rowN = matrix.size();
        int colN = matrix[0].size();
        
        DP.resize(rowN + 1, vector<int>(colN + 1, 0));
        // DP[i][j] = the sum of the i rows and first j columns
        
        for (int row = 0; row < rowN; row++) {
            for (int col = 0; col < colN; col++) {
                DP[row+1][col+1] = DP[row][col+1] + DP[row+1][col] + matrix[row][col] - DP[row][col];
            }
        }
    }
    
    int sumRegion(int row1, int col1, int row2, int col2) {
        return DP[row2+1][col2+1] - DP[row1][col2+1] - DP[row2+1][col1] + DP[row1][col1];
    }
};
````

### 306 Additive Number
 - Simplify the problem: The entire partitioning of the `num` string only depends on the length of the first two partitions
 - Integer to string `string::to_string(int)`
 - String to Integer `stoi(string)` if it is out-of-range `stol(string)`
 - When overflow, using `long` in place of `int`

````python
class Solution {
public:    
    bool checkRest(string & num, long begin, long num1, long num2) {
        if (begin >= num.size()) return false;
        
        while (begin < num.size()) {
            long num3 = num1 + num2;
            string str3 = to_string(num3);
            for (long i = 0; i < str3.size(); i++) {
                char c = str3[i];
                if (begin >= num.size()) return false;
                else if (c != num[begin]) return false;
                else begin++;
            }
            num1 = num2;
            num2 = num3;
        }
        return true;
    }
    
    bool isAdditiveNumber(string num) {
        // Handle degenerate cases
        if (num.size() < 3) {
            return false;
        }
        
        // Iterate through all possible (num1, num2) combinations
        long len1, len2;
        long rest_begin_index, num1, num2;
        bool found = false;
        
        for (len1 = 1; len1 <= num.size() / 2; len1++) {
            for (len2 = 1; len2 <= num.size() / 2; len2++) {
                if (len1 + len2  >= num.size()) {
                    break;
                } else if (num[0] == '0' && len1 > 1|| num[len1] == '0' && len2 > 1) {
                    break;
                } else {
                    rest_begin_index = num1 + num2;
                    num1 = stol(num.substr(0, len1));
                    num2 = stol(num.substr(len1, len2));
                    rest_begin_index = len1 + len2;
                    if (checkRest(num, rest_begin_index, num1, num2)) return true;
                }
            }
        }
        
        // Exit loop
        return false;
    }
};
````

### 307 Range Sum Query - Mutable

````python
class NumArray {
public:
    NumArray(vector<int>& nums) {
        int n = int(nums.size()); 
        
        prefixSum.resize(n+1);
        prefixSum[0] = 0;
        for (int i = 1; i <= n; i++) {
            prefixSum[i] = prefixSum[i-1] + nums[i-1];
        }
        
        for (int i = 0; i < n; i++) {
            numbers.push_back(nums[i]);
        }
    }
    
    void update(int index, int val) {
        int n = int(numbers.size()); 
        int offset = val - numbers[index];
        numbers[index] = val;
        
        for (int i = index+1; i <= n; i++) {
            prefixSum[i] += offset;
        }
    }
    
    int sumRange(int left, int right) {
        return prefixSum[right+1] - prefixSum[left];
    }
    
private:
    vector<int> numbers;   // the original numbers
    vector<int> prefixSum; // prefixSum[i] = the sum of the first i numbers
};
````

### 309 Best Time to Buy and Sell Stock with Cooldown
 > - Observation 1: The only factor for casing: `if the player holds stocks right now`
 > - Observation 2: To deal with the `cool down after selling` => `move two days instead of one day further when deciding to sell`
 - TopDown DP variables `DP(day, hasStock)`

````python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        cache = dict()
        
        # RETURNS: the maximum profit from `day` under the constraint that we `canBuy` on `day`
        # MODIFIES: store the result in the cache
        def DP(day, hasStock):
            if day >= len(prices): return 0
            if (day, hasStock) in cache: return cache[(day, hasStock)]
            
            if hasStock:
                # sell on this day
                sell = prices[day] + DP(day + 2, False)
                # cool down on this day
                cooldown = DP(day + 1, True)
                cache[(day, hasStock)] = max(sell, cooldown)
            else:
                # buy on  this day
                buy = (-prices[day]) + DP(day + 1, True)
                # cool down on this day
                cooldown = DP(day + 1, False)
                cache[(day, hasStock)] = max(buy, cooldown)
            
            return cache[(day, hasStock)]
        
        # Main
        res = DP(0, False)
        return res
````

### 310 Minimum Height Trees
 - 该函数可以用作 `Topological Degeneration by Frontier of Leaves` 的模板
 - Tip 1: Observe the implicit, underlying graph of the problem
 - 思路 : Onion Peal `At each round, pealing off all the leaves of the graph` 
 - Usually useful when hoping to find out the most `central, convex` point of a graph. Start from the external to the middle, advance frontier by frontier (of only leaves)
 - Wait untill the new leaves degenerate to be empty, the old leaves will be the second last frontier = the last non-empty frontier that remain to the last

````python
from collections import deque
class Solution:
    def findMinHeightTrees(self, n: int, edges: List[List[int]]) -> List[int]:
        # Handle degenerate cases
        if n == 1: return [0]
        if n == 2: return [0, 1]
        
        adj = [set() for i in range(n)]
        for edge in edges:
            u, v = edge[0], edge[1]
            adj[u].add(v)
            adj[v].add(u)
         
        new_leaves = [i for i in range(n) if len(adj[i]) == 1]
        old_leaves = [i for i in range(n) if len(adj[i]) == 1]
        
        # 
        while new_leaves:
            old_leaves = new_leaves.copy()
            new_leaves = []
            for l in old_leaves:
                for nbor_l in adj[l]:
                    adj[nbor_l].remove(l)
                    if len(adj[nbor_l]) == 1:
                        new_leaves.append(nbor_l)
                    
        return old_leaves
````

### 311 Sparse Matrix Multiplication
 - Normal matrix multiplication

````python
class Solution {
public:
    vector<vector<int>> multiply(vector<vector<int>>& mat1, vector<vector<int>>& mat2) {
        int m = mat1.size();
        int n = mat1[0].size();
        int l = mat2[0].size();
        
        vector<vector<int>>res(m, vector<int>(l, 0));
        
        for (int row = 0; row < m; row++) {
            for (int col = 0; col < l; col++) {
                // traverse the row-th row in mat1 and col-th col of mat2\
                
                for (int k = 0; k < n; k++) {
                    res[row][col] += mat1[row][k] * mat2[k][col];
                }
            }
        }
        
        return res;
    }
};
````

### 316 Remove Duplicate Letters
 - 思路: $\textit{We will keep a stack to store the solution we have built as we iterate over the string, and we will delete characters off the stack whenever it is possible and it makes our string smaller.}$
 - To check for the existance of a character in constant time -> `seen` and `sol` will contain the same content, but `seen` supports O(1) time check for existance of members
 - Note: subject to the constraint: we cannot change the order of the characters -> new characters that already exist in the `seen` has to be discarded
 - 使用 stack 的原因 -> add and remove from one end -> cannot change the order -> can only build the solution as we iterate from the left to the right


````python
class Solution {
public:
    string removeDuplicateLetters(string s) {
        
        stack<char> sol;            // the current solution so far
        set<char> seen;             // all distinct letters in the current solution
        map<char, int> last_index;  // mapping a character to its last inedx of occurance
        
        for (int i = 0; i < s.size(); i++) {
            char c = s[i];
            last_index[c] = i;
        }
        
        // invariant: at any time, the stack contains the subsequence from the prefix s[:i] that forms the best prefix
        // invariant: seen and sol contains the same set of chars, just in differen order, we need seen for O(1) check for existance
        for (int i = 0; i < s.size(); i++) {
            char c = s[i];
            if (seen.find(c) == seen.end()) {
                // (1) move as many characters off the top of the stack
                while (sol.size() > 0 && int(c) < int(sol.top()) && last_index[sol.top()] > i) {
                    seen.erase(sol.top());
                    sol.pop();
                }
                // (2) add the current character
                seen.insert(c);
                sol.push(c);
            }
        }
        
        cout << "here" << endl;
        
        string result = "";
        
        while (sol.size() > 0) {
            result = sol.top() + result;
            sol.pop();
        }
        
        return result;
    }
};
````

### 313 Super Ugly Number
 - Pattern / Situation: `the nth smallest number`
 - Greedy
 - Keep track of all the generated ugly numbers so far in a `priority_queue`
 - To prevent duplicate instances of ugly numbers + `O(1)` check -> use a set of all the elements present in the `priority_queue`

````python
class Solution {
public:
    int nthSuperUglyNumber(int n, vector<int>& primes) {
        set<long> generatedNumbers; // used to skip duplicate generate numbers
        priority_queue<long, vector<long>, greater<long>> PQ;
            
        // Initialize
        generatedNumbers.insert(1);
        PQ.push(1);
        
        long smallest;
        
        // Repeat for n time 
        for (long i = 0; i < n; i++) {
            smallest = PQ.top();
            PQ.pop();
            
            // generate the next possible smallest ugly number
            for (auto p : primes) {
                long newUgly = smallest * p;
               
                if (generatedNumbers.find(newUgly) == generatedNumbers.end()) {
                    generatedNumbers.insert(newUgly);
                    PQ.push(newUgly);
                }
            }
        }
        return smallest;
    }
};
````

### 320 Generalized Abbreviation
 - `Backtrack`
 - Keep track of `curAbbrev`, `cursor`, and a boolean flag for whether the previous part is the end of an abbreviation

````python
class Solution:
    def generateAbbreviations(self, word: str) -> List[str]:
        abbrevs = []
        
        def backtrack(cursor, curAbbrev, prevAbbrevEnd):
            # Base case
            if cursor >= len(word):
                abbrevs.append(curAbbrev)
                return
            
            # Recursive 
            if prevAbbrevEnd: # -> cannot abbrevite this one
                newAbbrev = curAbbrev + word[cursor]
                backtrack(cursor+1, newAbbrev, False)
            else:
                # -> option 1: do not abbreviate this one
                newAbbrev = curAbbrev + word[cursor]
                backtrack(cursor+1, newAbbrev, False)
                
                # -> option 2: begin abbreviation here
                # note: no need to extend upon the previou ongoing abbreivation
                for length in range(1, len(word)-cursor+1):
                    newAbbrev = curAbbrev + str(length)
                    backtrack(cursor+length, newAbbrev, True)
            
        # Driver
        backtrack(0, "", False)
        return abbrevs
````

### 322 Coin Change
- Initialize to be infinity
```
import math
math.inf
```
- Define `DP[n]` := the min number of coins used for amount = n
- Recurrece $DP[i] = \min_{target=i-coin:coin \in coins}1+DP[target]$

````python
import math
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        # Handle degeneraet case
        if amount == 0: return 0
        
        # Handle normal cases
        DP = [math.inf for _ in range(amount+1)] # DP[i] = min number of counts for amount of i
        
        # Base case
        DP[0] = 0
        
        # Recursive case
        for amt in range(1, amount+1):
            for coin in coins:
                if coin <= amt:
                    DP[amt] = min(DP[amt], 1 + DP[amt-coin])
        
        if DP[amount] == math.inf: return -1
        else: return DP[amount]
````

### 323 Number of Connected Components in an Undirected Graph
- DFS pseudocode
```
dfs(v):
    if v is visited:
        return
    else:
        mark v as visited
        for u in G.adj[v]:
            dfs(u)
```
- Write a helper function that covers an entire component beginning from `v` (implementation: by marking them as visited)
- Traverse all the vertices that are not visited, the number of `cover(v)` issued equals the number of components in the graph

````python
class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        # Construct the adjacency list
        adj = []
        for _ in range(n): adj.append([])
        for [u, v] in edges:
            adj[u].append(v)
            adj[v].append(u)
        
        visited = [False] * n
        
        def cover(v):
            visited[v] = True
            for nbor_v in adj[v]:
                if not visited[nbor_v]:
                    cover(nbor_v)
        
        components = 0
        for v in range(n):
            if not visited[v]:
                components += 1
                cover(v)
        return components
````

### 324 Wiggle Sort II
 - Sort the array:
 - Fill in the odd indices from max to min
 - Fill in the even indices from max to min

````python
class Solution {
public:
    void wiggleSort(vector<int>& nums) {    
        vector<int> sorted = nums;
        sort(sorted.begin(), sorted.end());
        
        int index = nums.size() - 1;
        
        // fill in odd indices from the max to min
        for (int i = 1; i < nums.size(); i+=2) 
            nums[i] = sorted[index--];
        
        // fill in odd indices from max to min
        for (int i = 0; i < nums.size(); i+=2) 
            nums[i] = sorted[index--];
    }
};
````

### 325 Maximum Size Subarray Sum Equals k
 - 思路 `PrefixSum Precomputation`  + `2Sum Complement Hashmap`
 - Step 1: Construct a prefixSum array `prefixSum[i] =` the sum of the first i elements of the array
 - Step 2: Iterate over all possible ending index of the potential array that sums to `k` -> exhaustive because the final answer array has to end some where iterated by `end` <br>
 And keep track of `prefixSum2End[s]` = the earlist index where a prefix array of sum `s` end at (has to be the earliest!)
    - (1) update the `prefixSum2End` only if `s not in prefixSum2End`
    - (2) compute the complement `prefixSumNeeded = prefixSumSoFar - k`
    - (3) Look for the corresponding earliest ending index for `prefixSumNeeded` if there is one.

````python
class Solution {
public:
    int maxSubArrayLen(vector<int>& nums, int k) {
        long n = nums.size();
        vector<long> prefixSum(n+1, 0);
        
        for (long l = 1; l <= n; l++) {
            prefixSum[l] = prefixSum[l-1] + nums[l-1];
        }
        
        map<long, long> prefixSum2End;
        prefixSum2End[0] = -1; // the prefix of sum 0 ends at index 0 inclusive
        long longest = 0;
        
        for (long end = 0; end < n; end++) {
            long prefixSumSoFar = prefixSum[end + 1];
            if (prefixSum2End.find(prefixSumSoFar) == prefixSum2End.end())
                prefixSum2End[prefixSumSoFar] = end;
            
            // prefixSumSoFar - prefixNeeded = k
            long prefixSumNeeded = prefixSumSoFar - k;
            
            if (prefixSum2End.find(prefixSumNeeded) != prefixSum2End.end()) {
                long start = prefixSum2End[prefixSumNeeded] + 1;
                long length = end - start + 1;
                longest = max(longest, length);
            }
        }
        
        return longest;
    }
};
````

### 329 Longest Increasing Path in a Matrix
 - Approach One: `DP[row][col]` = the length of the longest path, at most `mn` subproblems, each taking constant time $O(mn)$
    - Base case: with no larger neighbors -> already implicitly set to 0, eliminated by the conditional check `matrix[newRow][newCol] > matrix[row][col]`
 - Approcah Two: `Onion-peeling Algoirthm to get the diameter of a graph`
   - Diameter Finding Algorithm pseudo code 
   > Prerequisite: the graph must be acyclic. For this problem, the graph must be acyclic since we care about strictly increasing paths
      ```
         def getDiameter(adj, outdeg):
            leaves = [v for v in adj if outdeg[v] == 0]
            diameter = 0

            while leaves:
               diameter += 1
               newLeaves = []

               # Observation: new leaves can only come from neighbors of the current leaveas

               for v in leaves:
                  for nbor_v in adj[v]:
                     outdeg[nbor_v] -= 1
                     if outdeg[nbor_v] == 0:
                        newLeaves.append(nbor_v)
               
               leaves = newLeaves
            return diameter
      ```
   - The longest path is guaranteed by removing `ONLY` leaves from the graph and iterate untill there are no leaves remaining

````python
class Solution:
    def createMatrix(self, rowN, colN):
        return [ [1 for _ in range(colN)] for _ in range(rowN) ]
        
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        m = len(matrix)
        n = len(matrix[0])
        cache = self.createMatrix(m, n)
        
        # RETURNS: the length of the longest increasing path length beginning from [row][col]
        # MODIFIES: store the above result in the cache
        def dfs(row, col, m, n):
            # base case
            if cache[row][col] != 1:
                return cache[row][col]
            
            # recursive case
            for (dr, dc) in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                newRow = row + dr
                newCol = col + dc
                
                if 0 <= newRow < m and 0 <= newCol < n and matrix[newRow][newCol] > matrix[row][col]:
                    cache[row][col] = max(
                        cache[row][col],
                        1 + dfs(newRow, newCol, m, n)
                    )
            return cache[row][col]
        
        # DRIVER
        ans = 0
        for row in range(m):
            for col in range(n):
                ans = max(ans, dfs(row, col, m, n))
                
        print(cache)
        return ans

class Solution:
    def createMatrix(self, rowN, colN, init):
        return [ [ init for _ in range(colN)] for _ in range(rowN) ]
        
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        m = len(matrix)
        n = len(matrix[0])
        dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        
        # Count the out degreees of each cell
        outdegree = self.createMatrix(m, n, 0)
        
        for r in range(m):
            for c in range(n):
                for (dr, dc) in dirs:
                    newR, newC = r+dr, c+dc
                    if 0 <= newR < m and 0 <= newC < n and matrix[newR][newC] > matrix[r][c]:
                        outdegree[r][c] += 1
        
        
        def get_diameter(outdegree):
            # Collect all the leaves
            leaves = []
            for r in range(m):
                for c in range(n):
                    if outdegree[r][c] == 0:
                        leaves.append((r, c))

            # Remove leaves level by level in topological order
            diameter = 0
            while leaves:
                diameter += 1
                newLeaves = []
                # Observation: new leaves can only come from neighbors of the current leaves
                # disconnect the leaves from the current graph
                for (r, c) in leaves:
                    for (dr, dc) in dirs:
                        newR, newC = r+dr, c+dc
                        if 0 <= newR < m and 0 <= newC < n and matrix[newR][newC] < matrix[r][c]:
                            outdegree[newR][newC] -=1
                            if outdegree[newR][newC] == 0: 
                                newLeaves.append((newR, newC))
                leaves = newLeaves
            return diameter
        
        # Driver
        return get_diameter(outdegree)
            
````

### 331 Verify Preorder Serialization of a Binary Tree
 - Approach One: substring DP `DP[i, j]` = whether `preorder[i, j)` is a valid tree serialization
 - Approach Two: A spot can be used to place either a number of a null
    - If a number is used to fill a spot -> consume one spot -> check for deficit -> reward two spots
    - If a null is used to fill in a spot -> consume one spot -> check for deficit

````python
class Solution:
    def isValidSerialization(self, preorder: str) -> bool:
        preorder = preorder.split(',')
        
        spots = 1
        
        for node in preorder:
            if node == '#':
                if spots < 1: return False
                spots -= 1
            else:
                if spots < 1: return False
                spots += 1
        
        return spots == 0

class Solution:
    def isValidSerialization(self, preorder: str) -> bool:
        # Parse the string into a list of symbols
        preorder = preorder.split(',')
    
        # Handle degenerate case
        if len(preorder) == 0: return True
        if len(preorder) == 1: return preorder[0] == '#'
        
        # -> form here, an empty is not a valid tree
        
        # cache[i, j] whether preorder[i:j) makes a valid Binary Tree serialization
        cache = dict()
        
        # RETURNS: if preorder[i, j) i is a valid serialization
        # MODIFIES: store the result in the cache
        def DP(i, j):
            # Base case:
            if (i, j) in cache:
                return cache[(i, j)]
            if j - i <= 0 or j - i == 2: 
                cache[(i, j)] = False
                return False
            if j - i == 1: 
                cache[(i, j)] = preorder[i] == '#'
                return preorder[i] == '#'
            if preorder[i] == '#':
                cache[(i, j)] = False
                return False
            
            # Recursive case
            # iterate over all possible length of the left subtree of preorder[0]
            minLeftSize = 1
            maxLeftSize = j - i - 2

            for leftSize in range(minLeftSize, maxLeftSize+1):
                leftValid = DP(i+1, i+1+leftSize)
                rightValid = DP(i+1+leftSize, j)
                if leftValid and rightValid:
                    cache[(i, j)] = True
                    return True
            cache[(i, j)] = False
            return False
    
        # DRIVER
        DP(0, len(preorder))
        return cache[(0, len(preorder))]
````

### 332 Reconstruct Itinerary
 - 思路: Model the tickets into a graph. In order to construct the lexically smallest itenary, the entries in the `adjacency list` must be lexically sorted. To do this, just sort all the tickets: first according to the `src`, if `src` is the same,  the compare according to `dest`. Traversing the sorted list to build `adj` ensures that each value will be sorted.
 Since the traveller can arrive at an airport multiple times, we do not need to care about `cycle, stack, visited`. Instead, each time we make a travel through an edge `u -> v`, just do the following
   1. delete `v` from `adj[u]` 
       - Implemented by `popleft` from a `deque`
   2. Recursive search
   3. add `v` back to `adj[u]` when this round fails
       - Implemented by `append` to a `deque`

````python
 class Solution:
    def findItinerary(self, tickets: List[List[str]]) -> List[str]:
        # sort the tickets first based on src and then based on dest
        # -> to make each list in the codomain of adj sorted
        tickets.sort(key = lambda ticket: (ticket[0], ticket[1]))
        
        # construct the adjacency list
        adj = dict() 
        for src, _ in tickets:
            adj[src] = deque([])
        for src, dest in tickets:
            adj[src].append(dest)
        
        itenary = ["JFK"]
        
        # start searching the graph from v (assuming v has already been added to the path)
        # RETURNS: if starting from v there is a valid itenary
        # MODIFIES: the `itenary`
        def DFS(v):
            # Base case
            if len(itenary) == len(tickets) + 1:
                return True
            if v not in adj or not adj[v]:
                return False
            
            # Recursive case
            destCount = len(adj[v])
            for _ in range(destCount):
                dest = adj[v].popleft()
                itenary.append(dest)
                
                if DFS(dest): return True
                
                itenary.pop()
                adj[v].append(dest)
            
            return False
        
        # Driver
        DFS("JFK")
        return itenary
````

### 334 Increasing Triplet Subsequence
 - Problem: find i < j < k such that A[i] < A[j] < A[k]
 - The actual existence of `k` only depends on the value of `A[j]`. Greedily, keep `j_current` keeps track of the currently achievable minimum such`A[j]`. This means that there always exists a value (ceonceptually the CURRENT `A[i]`) that is both (1) smaller and (2) comes before it.
 - `i_backup` keeps tack of the NEXT potential `A[i]`.

````python
import math
class Solution:
    # greedy
    def increasingTriplet(self, nums: List[int]) -> bool:
        if len(nums) < 2:
            return False
        
        i_backup = nums[0] 
        j_current = math.inf
        # implicity, i_backup < j_current
        
        for num in nums[1:]:
            if num <= i_backup:
                i_backup = num
            elif num <= j_current:
                j_current = num
            else:
                return True
            # print('num', num, 'i_backup', i_backup, 'j_current', j_current)
        return False
````

### 337 House Robber III
- Keep track of two sets of DP caches
- `DP(root, robHere)` means if we do rob here or not, not if we can rob here or not.
- O(n)

````python
class Solution:
    def rob(self, root: Optional[TreeNode]) -> int:
        robbed = {None: 0}
        skipped = {None: 0}
        
        def DP(root, robHere):
            # Base case
            if root is None: return 0
            if robHere and root in robbed: return robbed[root]
            if not robHere and root in skipped: return skipped[root]
            
            # Recursive case
            if robHere:
                gain = DP(root.left, False) + DP(root.right, False) + root.val
                robbed[root] = gain
                return gain
            else:
                robBothChild = DP(root.left, True) + DP(root.right, True)
                robLeftChild = DP(root.left, True) + DP(root.right, False)
                robRightChild = DP(root.left, False) + DP(root.right, True)
                robNoChild = DP(root.left, False) + DP(root.right, False)
                gain = max(
                    robBothChild,
                    robLeftChild,
                    robRightChild,
                    robNoChild
                )
                skipped[root] = gain
                return gain
        
        robRoot = DP(root, True)
        skipRoot = DP(root, False)
        return max(robRoot, skipRoot)
````

### 340 Longest Substring with At Most K Distinct Characters
 - At most k distinct characters
 - `cnt` = number of distinct characters
 - `freq[c]` = number of its occurence in the current window 
 - No need to add a conditional in step 3 since we are always able to restore the property in step 2

````python
class Solution:
    def lengthOfLongestSubstringKDistinct(self, s: str, k: int) -> int:
        # handle degenerate cases
        if k == 0:
            return 0
        
        # Handle normal cases
        begin, end, cnt, freq = 0, 0, 0, defaultdict(int) # initialized to be a valid window
        length = 0
        
        while end < len(s):
            # move the `end` ptr to the right by one
            endChar = s[end]
            freq[endChar] += 1
            if freq[endChar] == 1: # a new character has been included in the window
                cnt += 1
            end += 1

            # keep moving `end` to the right to restore the window
            while cnt > k:
                beginChar = s[begin]
                freq[beginChar] -= 1
                if freq[beginChar] == 0:
                    cnt -= 1
                begin += 1
            
            # now that the window [begin, end) is good is valid 
            length = max(length, end-begin)
            
        return length
````

### 345 Reverse Vowels of a String
 - Simple `two pointer`
 - `left` -> the index of the un-traversed, left most, un-reversed vowel from the left
 - `right` -> the index of the un-traversed, right most, un-reversed vowel from the right

````python
class Solution {
public:
    const set<char> vowels{'a', 'e', 'i', 'o', 'u',
                           'A', 'E', 'I', 'O', 'U'};
    
    bool is_vowel(char c) {
        return vowels.find(c) != vowels.end();    
    }
    
    string reverseVowels(string s) {
        int left, right, n = s.size();
        
        // initialize left to the left most vowel
        left = 0;
        while (left < n and !is_vowel(s[left])) left++;
        
        
        // initialize right to the right most vowel
        right = n-1;
        while (right >= 0 and !is_vowel(s[right])) right--;
        
        while (left < right) {
            char tmp = s[left];
            s[left] = s[right];
            s[right] = tmp;
            
            // move left and right to the next state
            left++;
            while (left < n and !is_vowel(s[left])) left++;
            right--;
            while (right >= 0 and !is_vowel(s[right])) right--;
        }
        
        return s;
    }
};
````

### 346 Moving Average from Data Stream
 - A queue data structure $\textit{Keep track of the n most recent elements}$
 - Keep track of a queue `Q`
 - When there are enough elements
    - pop the least recent from the front of the queue
    - push the new val to the back of the queue
    - add new val to the cumulative sum
    - subtract the poped tail from the cumulative sum

````python
class MovingAverage {
public:
    int window_size, cur_size;
    double cur_sum;
    queue<double> Q;
    
    MovingAverage(int size) {
        window_size = size;
        cur_size = 0;
        cur_sum = 0.0;
    }
    
    double next(int val) {
        cur_size++;
        // cout << "cur sum " << cur_sum << " cur size " << cur_size << endl;
        
        if (cur_size <= window_size) {
            Q.push(val);
            cur_sum += val;
            return cur_sum / cur_size;
        } else {
            double tail = Q.front();
            Q.pop();
            cur_sum += (val - tail);
            Q.push(val);
            return cur_sum / window_size;
        }
    }
};
````

### 347 Top K Frequent Elements
 - Keep track of a counter and then traverse the counter `->` always make sure the heap size `<=k`, so it is only necessary to compare the iterated (key, value) pair with the top of the heap
 - Cannot pass multiple arguements to a `lambda` expression -> `lambda pair : pair[1]`

````python
import heapq

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        freq = collections.Counter(nums)
        
        PQ = []
        heapq.heapify(PQ)
        
        for num in freq:
            if len(PQ) < k:
                heapq.heappush(PQ, (freq[num], num))
            else:
                (topFreq, topElement) = heapq.heappop(PQ)
                if freq[num] < topFreq:
                    heapq.heappush(PQ, (topFreq, topElement))
                else:
                    heapq.heappush(PQ, (freq[num], num))
        
        return map(lambda pair : pair[1], list(PQ))
                
````

### 350 Intersection of Two Arrays II
 - Use hashmap `freq = Counter(nums)`
 - Swap `nums1` and `nums2` to make sure `nums1` is always pointing to the shorter one

````python
def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
    # Swap so that num1 is the shorter array
    if len(nums1) > len(nums2):
        tmp = nums1
        nums1 = nums2
        nums2 = tmp
    
    freq1 = Counter(nums1)
    freq2 = Counter(nums2)
    
    intersection = []
    
    for key in freq1:
        if key in freq2:
            intersection = intersection + [key] * (min(freq1[key], freq2[key]))
    return intersection
````

### 355 Design Twitter

````python
from collections import defaultdict

class Twitter:

    def __init__(self):
        self.tweets = defaultdict(list)   # userID => list of (time, tweetID)
        self.followee = defaultdict(set)    # userID => set of userID
        self.clock = 0

    def postTweet(self, userId: int, tweetId: int) -> None:
        self.followee[userId].add(userId)
        self.tweets[userId].append((self.clock, tweetId))
        self.clock -= 1

    def getNewsFeed(self, userId: int) -> List[int]:
        self.followee[userId].add(userId)
        minHeap = []
        heapq.heapify(minHeap)
        
        # get a handler to all the followees, including user itself
        for followedId in self.followee[userId]:
            if followedId in self.tweets:
                latestTweetIndex = len(self.tweets[followedId]) - 1
                latestTime, latestTweetId = self.tweets[followedId][latestTweetIndex]
                # info contained: [followedId] sent the [latestTweetId] at the time of [latestTime]
                heapq.heappush(minHeap, (latestTime, latestTweetId, latestTweetIndex, followedId))
        
        # collect the 10 latest tweetIds
        feed = []
        while len(feed) < 10 and len(minHeap) > 0:
            latestTime, latestTweetId, latestIndex, latestFollowed = heapq.heappop(minHeap)
            feed.append(latestTweetId)
            # add the next latest tweet info of this handler into the heap
            if latestIndex > 0:
                nextTime, nextTweetId = self.tweets[latestFollowed][latestIndex - 1]
                heapq.heappush(minHeap, (nextTime, nextTweetId, latestIndex-1, latestFollowed))
        return feed

    def follow(self, followerId: int, followeeId: int) -> None:
        self.followee[followerId].add(followeeId)

    def unfollow(self, followerId: int, followeeId: int) -> None:
        if followeeId in self.followee[followerId]:
            self.followee[followerId].remove(followeeId)


# Your Twitter object will be instantiated and called as such:
# obj = Twitter()
# obj.postTweet(userId,tweetId)
# param_2 = obj.getNewsFeed(userId)
# obj.follow(followerId,followeeId)
# obj.unfollow(followerId,followeeId)
````

### 371 Sum of Two Integers
 - 思路 : Normal decimal vertical-form addition also works for binary vertical-form arithmetic. Thinking in terms of the vertical form, try to simulate its behavior using bit-wise operators.
   - Coordinate-wise addition: `bitwise-xor`
   - Carry vector: `bitwise-and`
   - Final result: `bitwise-or` 
 - The last step must not have a carry !

````python
class Solution:    
    def getSum(self, a: int, b: int) -> int:        
        withoutCarry = a ^ b
        carry = a & b 
        res = withoutCarry | (carry << 1)
        
        return res
````

### 378 Kth Smallest Element in a Sorted Matrix
 - $Kth$ 问题一般采用 Priority Priority Queue
 - Keep track of the smallest boundary
 ```
  *  *  * |
  *  |
  *  *  *  *  *  |
  *  *  *  *  *  |
 ```
 - Only need to keep track of the $min(M, k)$ rows and push the sentinels of each row forward
 - $invariant$ 每次iteration中pop出的值一定是最小的
 - 记录第k个被poped的值
 - `heapify(list of length X)` costs $O(X)$

````python
import heapq
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        M = len(matrix) # number of rows
        N = len(matrix[0]) # number of cols
        PQ = []
        for i in range(min(M, k)):
            key = matrix[i][0]
            value = (i, 0)
            PQ.append((key, value))
        
        heapq.heapify(PQ)
        lastPoped = None
        
        for _ in range(k):
            # pop the current min 
            (num, (r, c)) = heapq.heappop(PQ)
            lastPoped = num
            
            # push the current min into heap
            if c+1 < M:
                newKey = matrix[r][c+1]
                newVal = (r, c+1)
                heapq.heappush(PQ, (newKey, newVal))
        
        return lastPoped
        # costs O(X) + O(klogX)
````

### 394 Decode String
- `Stack` data structure
```
S = []
S.appennd(elem)
elem = S.pop()
```
- Reverse a string `s[::-1]`
- Join a list of strings into one `"".join(strs)`

````python
class Solution:
    def decodeString(self, s: str) -> str:
        n = len(s)
        stack = []
        i = n-1
        
        while i >= 0:
            token = s[i]

            if token.isdigit():
                repeat = ""

                # (1) read the integer of how many times to repeat
                while token.isdigit():
                    repeat = token + repeat
                    i -= 1
                    token = s[i]

                # (2) read the string to be repeated
                stack.pop()
                delta_s = ""
                while stack[-1] != ']':
                    delta_s += stack.pop()
                stack.pop()
                
                # (3) push the repeated string back to stack
                stack.append(int(repeat) * delta_s)
            else:
                stack.append(token)
                i -= 1
        
        stack.reverse()
        return "".join(stack)
````
