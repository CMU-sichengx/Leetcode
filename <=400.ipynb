### 302 Smallest Rectangle Enclosing Black Pixels
 - Approach 1 : `BFS` the black region and update the `top`, `buttom`, `left`, and `right` when visiting each cell
     - O(mn) times for BFS
 - Approcah 2: Binary search `2(MlogN + NlogM)`
     - Binary search `N` columns each taking `logM` to find left
     - Do the same for right
     - Binary search `M` rows each taking `logN` to find top
     - Do the same for buttom 

````python
class Solution {
public:
    int top, buttom, left, right;
    
    void dfs(vector<vector<char>>& image, int x, int y){
        // Base case
        if (x < 0 || x >= image.size() || y < 0 || y >= image[0].size() || image[x][y] == '0') return;
        
        // Recursive case
        image[x][y] = '0';
        top = min(top, x);
        buttom = max(buttom, x);
        left = min(left, y);
        right = max(right, y);
        
        dfs(image, x, y-1);
        dfs(image, x, y+1);
        dfs(image, x-1, y);
        dfs(image, x+1, y);
    }
    int minArea(vector<vector<char>>& image, int x, int y) {
        left = y;
        right = y;
        top = x;
        buttom = x;
        dfs(image, x, y);
        cout << left << " " << right << " " << top << " " << buttom << endl;
        return (right + 1 - left) * (buttom + 1 - top);
    }
};
````

### 303 Range Sum Query - Immutable
 - Prefix sum method
 - A *contiguous reduce sum of an array* -> *prefix sum*


````python
class NumArray:

def __init__(self, nums: List[int]):
    n = len(nums)
    prefSum = [0] * (n+1) # prefSum[i] = the sum of the first i numbers of nums
    
    for i in range(1, n+1):
        prefSum[i] = prefSum[i-1] + nums[i-1]
    
    self.prefSum = prefSum
    

def sumRange(self, left: int, right: int) -> int:
    return self.prefSum[right+1] - self.prefSum[left]
````

### 304 Range Sum Query 2D - Immutable
 - The 2D version of the prfix sum
 - `DP[rowN][colN]` = the sum of the first rowN rows and the first colN cols
 - Recurrence $DP[rowN][colN] = DP[rowN-1][colN] + DP[rowN][colN-1] + matrix[rowN-1][colN-1]$

````python
class NumMatrix {
public:
    vector<vector<int>> DP; 
    NumMatrix(vector<vector<int>>& matrix) {
        int rowN = matrix.size();
        int colN = matrix[0].size();
        
        DP.resize(rowN + 1, vector<int>(colN + 1, 0));
        // DP[i][j] = the sum of the i rows and first j columns
        
        for (int row = 0; row < rowN; row++) {
            for (int col = 0; col < colN; col++) {
                DP[row+1][col+1] = DP[row][col+1] + DP[row+1][col] + matrix[row][col] - DP[row][col];
            }
        }
    }
    
    int sumRegion(int row1, int col1, int row2, int col2) {
        return DP[row2+1][col2+1] - DP[row1][col2+1] - DP[row2+1][col1] + DP[row1][col1];
    }
};
````

### 306 Additive Number
 - Simplify the problem: The entire partitioning of the `num` string only depends on the length of the first two partitions
 - Integer to string `string::to_string(int)`
 - String to Integer `stoi(string)` if it is out-of-range `stol(string)`
 - When overflow, using `long` in place of `int`

````python
class Solution {
public:    
    bool checkRest(string & num, long begin, long num1, long num2) {
        if (begin >= num.size()) return false;
        
        while (begin < num.size()) {
            long num3 = num1 + num2;
            string str3 = to_string(num3);
            for (long i = 0; i < str3.size(); i++) {
                char c = str3[i];
                if (begin >= num.size()) return false;
                else if (c != num[begin]) return false;
                else begin++;
            }
            num1 = num2;
            num2 = num3;
        }
        return true;
    }
    
    bool isAdditiveNumber(string num) {
        // Handle degenerate cases
        if (num.size() < 3) {
            return false;
        }
        
        // Iterate through all possible (num1, num2) combinations
        long len1, len2;
        long rest_begin_index, num1, num2;
        bool found = false;
        
        for (len1 = 1; len1 <= num.size() / 2; len1++) {
            for (len2 = 1; len2 <= num.size() / 2; len2++) {
                if (len1 + len2  >= num.size()) {
                    break;
                } else if (num[0] == '0' && len1 > 1|| num[len1] == '0' && len2 > 1) {
                    break;
                } else {
                    rest_begin_index = num1 + num2;
                    num1 = stol(num.substr(0, len1));
                    num2 = stol(num.substr(len1, len2));
                    rest_begin_index = len1 + len2;
                    if (checkRest(num, rest_begin_index, num1, num2)) return true;
                }
            }
        }
        
        // Exit loop
        return false;
    }
};
````

### 307 Range Sum Query - Mutable
 - Use prefix Sum `Time Limit Exceed`

````python
class NumArray {
public:
    NumArray(vector<int>& nums) {
        int n = int(nums.size()); 
        
        prefixSum.resize(n+1);
        prefixSum[0] = 0;
        for (int i = 1; i <= n; i++) {
            prefixSum[i] = prefixSum[i-1] + nums[i-1];
        }
        
        for (int i = 0; i < n; i++) {
            numbers.push_back(nums[i]);
        }
    }
    
    void update(int index, int val) {
        int n = int(numbers.size()); 
        int offset = val - numbers[index];
        numbers[index] = val;
        
        for (int i = index+1; i <= n; i++) {
            prefixSum[i] += offset;
        }
    }
    
    int sumRange(int left, int right) {
        return prefixSum[right+1] - prefixSum[left];
    }
    
private:
    vector<int> numbers;   // the original numbers
    vector<int> prefixSum; // prefixSum[i] = the sum of the first i numbers
};
````

### 311 Sparse Matrix Multiplication
 - Normal matrix multiplication

````python
class Solution {
public:
    vector<vector<int>> multiply(vector<vector<int>>& mat1, vector<vector<int>>& mat2) {
        int m = mat1.size();
        int n = mat1[0].size();
        int l = mat2[0].size();
        
        vector<vector<int>>res(m, vector<int>(l, 0));
        
        for (int row = 0; row < m; row++) {
            for (int col = 0; col < l; col++) {
                // traverse the row-th row in mat1 and col-th col of mat2\
                
                for (int k = 0; k < n; k++) {
                    res[row][col] += mat1[row][k] * mat2[k][col];
                }
            }
        }
        
        return res;
    }
};
````

### 316 Remove Duplicate Letters
 - 思路: $\textit{We will keep a stack to store the solution we have built as we iterate over the string, and we will delete characters off the stack whenever it is possible and it makes our string smaller.}$
 - To check for the existance of a character in constant time -> `seen` and `sol` will contain the same content, but `seen` supports O(1) time check for existance of members
 - Note: subject to the constraint: we cannot change the order of the characters -> new characters that already exist in the `seen` has to be discarded
 - 使用 stack 的原因 -> add and remove from one end -> cannot change the order -> can only build the solution as we iterate from the left to the right


````python
class Solution {
public:
    string removeDuplicateLetters(string s) {
        
        stack<char> sol;            // the current solution so far
        set<char> seen;             // all distinct letters in the current solution
        map<char, int> last_index;  // mapping a character to its last inedx of occurance
        
        for (int i = 0; i < s.size(); i++) {
            char c = s[i];
            last_index[c] = i;
        }
        
        // invariant: at any time, the stack contains the subsequence from the prefix s[:i] that forms the best prefix
        // invariant: seen and sol contains the same set of chars, just in differen order, we need seen for O(1) check for existance
        for (int i = 0; i < s.size(); i++) {
            char c = s[i];
            if (seen.find(c) == seen.end()) {
                // (1) move as many characters off the top of the stack
                while (sol.size() > 0 && int(c) < int(sol.top()) && last_index[sol.top()] > i) {
                    seen.erase(sol.top());
                    sol.pop();
                }
                // (2) add the current character
                seen.insert(c);
                sol.push(c);
            }
        }
        
        cout << "here" << endl;
        
        string result = "";
        
        while (sol.size() > 0) {
            result = sol.top() + result;
            sol.pop();
        }
        
        return result;
    }
};
````

### 313 Super Ugly Number
 - Pattern / Situation: `the nth smallest number`
 - Greedy
 - Keep track of all the generated ugly numbers so far in a `priority_queue`
 - To prevent duplicate instances of ugly numbers + `O(1)` check -> use a set of all the elements present in the `priority_queue`

````python
class Solution {
public:
    int nthSuperUglyNumber(int n, vector<int>& primes) {
        set<long> generatedNumbers; // used to skip duplicate generate numbers
        priority_queue<long, vector<long>, greater<long>> PQ;
            
        // Initialize
        generatedNumbers.insert(1);
        PQ.push(1);
        
        long smallest;
        
        // Repeat for n time 
        for (long i = 0; i < n; i++) {
            smallest = PQ.top();
            PQ.pop();
            
            // generate the next possible smallest ugly number
            for (auto p : primes) {
                long newUgly = smallest * p;
               
                if (generatedNumbers.find(newUgly) == generatedNumbers.end()) {
                    generatedNumbers.insert(newUgly);
                    PQ.push(newUgly);
                }
            }
        }
        return smallest;
    }
};
````

### 320 Generalized Abbreviation
 - `Backtrack`
 - Keep track of `curAbbrev`, `cursor`.
 - Case 1: `curAbbrev` is empty OR `curAbbrev` ends with a English letter
    - option1: start a new abbreivation with `1`
    - option2: do not abbreivate and keep the English character
 - Case 2: `curAbbrev` ends with a digit
    - option1: extend the current abbreivation, add one to the previous length
    - option2: do not abbreivate and keep the English character

````python
class Solution {
public:
    vector<string> abbrevs;
    
    // prevAbbreviated: whenther [cursor-1] is the end of an abbreviation
    void backtrack(string word, int cursor, string curAbbrev) {
        // base case
        if (cursor >= word.size()) {
            abbrevs.push_back(curAbbrev);
            return;
        }
        
        // recursive case
        if (curAbbrev.size() == 0 || curAbbrev.size() > 0 && !isdigit(curAbbrev.back())) {
            // option1: do not abbreviate
            curAbbrev.push_back(word[cursor]);
            backtrack(word, cursor+1, curAbbrev);
            curAbbrev.pop_back();
            
            // option2: start a new abbreivation
            curAbbrev.push_back('1');
            backtrack(word, cursor+1, curAbbrev);
            curAbbrev.pop_back();
            
        } else {
            // option1: do not abbreviate
            curAbbrev.push_back(word[cursor]);
            backtrack(word, cursor+1, curAbbrev);
            curAbbrev.pop_back();
            
            // option2: extend the ongoing abbreivation
            int curLength = int(curAbbrev.back());
            
            if (curAbbrev.size() >= 2 and isdigit(curAbbrev[curAbbrev.size()-2])) {
                curLength = 10 * int(curAbbrev[curAbbrev.size()-2]) + curLength;
            }
            
            int nextLength = curLength + 1;
            curAbbrev.pop_back();
            curAbbrev.push_back(char(nextLength));
            backtrack(word, cursor+1, curAbbrev);
            curAbbrev.pop_back();
        } 
    }    
    
    vector<string> generateAbbreviations(string word) {
        string curAbbrev = "";
        backtrack(word, 0, curAbbrev);
        return abbrevs;
    }
};
````

### 322 Coin Change
- Initialize to be infinity
```
import math
math.inf
```
- Define `DP[n]` := the min number of coins used for amount = n
- Recurrece $DP[i] = \min_{target=i-coin:coin \in coins}1+DP[target]$

````python
import math
class Solution:
    # complecity O(n * S)
    def coinChange(self, coins: List[int], amount: int) -> int:
        # Handle degenerate case
        if amount == 0:
            return 0
        if amount == 1:
            if 1 in coins:
                return 1
            else:
                return -1
        # DP[n] := the min number of coins used for amount = n
        DP = [math.inf] * (amount + 1)
        
        # Base Case 
        DP[0] = 0
        DP[1] = 1 if 1 in coins else math.inf
        
        # Recursive Case
        for i in range(2, amount+1):
            for coin in coins:
                target = i - coin
                if target >= 0:
                    DP[i] = min(DP[i], 1+DP[target])
        if DP[amount] == math.inf:
            return -1
        else:
            return DP[amount]

    best_sol = math.inf
    
    def back_track(cursor, remain, cur_sol):
        # base case
        if remain == 0:
            best_sol = min(best_sol, cur_sol)
            return;
        if remain < 0:
            return
        # recursive case
        for new_cursor in range(cursor, len(coins)):
            coin = coins[new_cursor]
            if coin <= remain:
                back_track(new_cursor, remain-coin, cur_sol+1)
    return best_sol
````

### 323 Number of Connected Components in an Undirected Graph
- DFS pseudocode
```
dfs(v):
    if v is visited:
        return
    else:
        mark v as visited
        for u in G.adj[v]:
            dfs(u)
```

````python
class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        # global state variables
        # 1
        visited = set()
        # 2
        nodeToNbor = []
        for _ in range(n):
            nodeToNbor.append([])

        for [u, v] in edges:
            nodeToNbor[u].append(v)
            nodeToNbor[v].append(u)
        # 3
        compN = 0
        
        # Define helper functions
        # dfs covering the component induced by "node"
        def dfs(node):
            # Base Case
            if node in visited:
                return
            # Recursive Case:
            visited.add(node)
            for nbor in nodeToNbor[node]:
                dfs(nbor)
            return
        
        # Main
        for node in range(n):
            if node not in visited:
                compN += 1
                dfs(node)
        return compN
````

### 324 Wiggle Sort II

 - Fill in the odd indices from max to min
 - Fill in the even indices from max to min

````python
class Solution {
public:
    void wiggleSort(vector<int>& nums) {    
        vector<int> sorted = nums;
        sort(sorted.begin(), sorted.end());
        
        int index = nums.size() - 1;
        
        // fill in odd indices from the max to min
        for (int i = 1; i < nums.size(); i+=2) 
            nums[i] = sorted[index--];
        
        // fill in odd indices from max to min
        for (int i = 0; i < nums.size(); i+=2) 
            nums[i] = sorted[index--];
    }
};
````

### 334 Increasing Triplet Subsequence
 - Problem: find i < j < k such that A[i] < A[j] < A[k]
 - The actual existence of `k` only depends on the value of `A[j]`. Greedily, keep `j_current` keeps track of the currently achievable minimum such`A[j]`. This means that there always exists a value (ceonceptually the CURRENT `A[i]`) that is both (1) smaller and (2) comes before it.
 - `i_backup` keeps tack of the NEXT potential `A[i]`.

````python
import math
class Solution:
    # greedy
    def increasingTriplet(self, nums: List[int]) -> bool:
        if len(nums) < 2:
            return False
        
        i_backup = nums[0] 
        j_current = math.inf
        # implicity, i_backup < j_current
        
        for num in nums[1:]:
            if num <= i_backup:
                i_backup = num
            elif num <= j_current:
                j_current = num
            else:
                return True
            # print('num', num, 'i_backup', i_backup, 'j_current', j_current)
        return False
````

### 338 Counting Bits
- Recurrence $DP[i] = DP[i//2] + (i \mod 2)$

````python
class Solution:
    def countBits(self, n: int) -> List[int]:
        # Handle degenerate cases
        if n == 0:
            return [0]
        if n == 1:
            return [0,1] 
        # DP[i] := popcount of i
        DP = [0] * (n+1)
        # Base Case
        DP[0] = 0
        DP[1] = 1
        # Recursive Case
        for i in range(2, n+1):
            DP[i] = DP[i>>1] + (i & 1)
        return DP
````

### 340 Longest Substring with At Most K Distinct Characters
 - At most k distinct characters
 - `cnt` = number of distinct characters
 - `freq[c]` = number of its occurence in the current window 
 - No need to add a conditional in step 3 since we are always able to restore the property in step 2

````python
class Solution:
    def lengthOfLongestSubstringKDistinct(self, s: str, k: int) -> int:
        # handle degenerate cases
        if k == 0:
            return 0
        
        # Handle normal cases
        begin, end, cnt, freq = 0, 0, 0, defaultdict(int) # initialized to be a valid window
        length = 0
        
        while end < len(s):
            # move the `end` ptr to the right by one
            endChar = s[end]
            freq[endChar] += 1
            if freq[endChar] == 1: # a new character has been included in the window
                cnt += 1
            end += 1

            # keep moving `end` to the right to restore the window
            while cnt > k:
                beginChar = s[begin]
                freq[beginChar] -= 1
                if freq[beginChar] == 0:
                    cnt -= 1
                begin += 1
            
            # now that the window [begin, end) is good is valid 
            length = max(length, end-begin)
            
        return length
````

### 345 Reverse Vowels of a String
 - Simple `two pointer`
 - `left` -> the index of the un-traversed, left most, un-reversed vowel from the left
 - `right` -> the index of the un-traversed, right most, un-reversed vowel from the right

````python
class Solution {
public:
    const set<char> vowels{'a', 'e', 'i', 'o', 'u',
                           'A', 'E', 'I', 'O', 'U'};
    
    bool is_vowel(char c) {
        return vowels.find(c) != vowels.end();    
    }
    
    string reverseVowels(string s) {
        int left, right, n = s.size();
        
        // initialize left to the left most vowel
        left = 0;
        while (left < n and !is_vowel(s[left])) left++;
        
        
        // initialize right to the right most vowel
        right = n-1;
        while (right >= 0 and !is_vowel(s[right])) right--;
        
        while (left < right) {
            char tmp = s[left];
            s[left] = s[right];
            s[right] = tmp;
            
            // move left and right to the next state
            left++;
            while (left < n and !is_vowel(s[left])) left++;
            right--;
            while (right >= 0 and !is_vowel(s[right])) right--;
        }
        
        return s;
    }
};
````

### 346 Moving Average from Data Stream
 - A queue data structure $\textit{Keep track of the n most recent elements}$
 - Keep track of a queue `Q`
 - When there are enough elements
    - pop the least recent from the front of the queue
    - push the new val to the back of the queue
    - add new val to the cumulative sum
    - subtract the poped tail from the cumulative sum

````python
class MovingAverage {
public:
    int window_size, cur_size;
    double cur_sum;
    queue<double> Q;
    
    MovingAverage(int size) {
        window_size = size;
        cur_size = 0;
        cur_sum = 0.0;
    }
    
    double next(int val) {
        cur_size++;
        // cout << "cur sum " << cur_sum << " cur size " << cur_size << endl;
        
        if (cur_size <= window_size) {
            Q.push(val);
            cur_sum += val;
            return cur_sum / cur_size;
        } else {
            double tail = Q.front();
            Q.pop();
            cur_sum += (val - tail);
            Q.push(val);
            return cur_sum / window_size;
        }
    }
};
````

### 350 Intersection of Two Arrays II
 - Use hashmap `freq = Counter(nums)`
 - Swap `nums1` and `nums2` to make sure `nums1` is always pointing to the shorter one

````python
def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
    # Swap so that num1 is the shorter array
    if len(nums1) > len(nums2):
        tmp = nums1
        nums1 = nums2
        nums2 = tmp
    
    freq1 = Counter(nums1)
    freq2 = Counter(nums2)
    
    intersection = []
    
    for key in freq1:
        if key in freq2:
            intersection = intersection + [key] * (min(freq1[key], freq2[key]))
    return intersection
````

### 378 Kth Smallest Element in a Sorted Matrix
 - $Kth$ 问题一般采用 Priority Priority Queue
 - Keep track of the smallest boundary
 ```
  *  *  * |
  *  |
  *  *  *  *  *  |
  *  *  *  *  *  |
 ```
 - Only need to keep track of the $min(M, k)$ rows and push the sentinels of each row forward
 - $invariant$ 每次iteration中pop出的值一定是最小的
 - 记录第k个被poped的值
 - `heapify(list of length X)` costs $O(X)$

````python
import heapq
class Solution:
    def kthSmallest(self, matrix: List[List[int]], k: int) -> int:
        M = len(matrix) # number of rows
        N = len(matrix[0]) # number of cols
        PQ = []
        for i in range(min(M, k)):
            key = matrix[i][0]
            value = (i, 0)
            PQ.append((key, value))
        
        heapq.heapify(PQ)
        lastPoped = None
        
        for _ in range(k):
            # pop the current min 
            (num, (r, c)) = heapq.heappop(PQ)
            lastPoped = num
            
            # push the current min into heap
            if c+1 < M:
                newKey = matrix[r][c+1]
                newVal = (r, c+1)
                heapq.heappush(PQ, (newKey, newVal))
        
        return lastPoped
        # costs O(X) + O(klogX)
````

### 394 Decode String
- `Stack` data structure
```
S = []
S.appennd(elem)
elem = S.pop()
```
- Reverse a string `s[::-1]`
- Join a list of strings into one `"".join(strs)`

````python
class Solution:
    def decodeString(self, s: str) -> str:
        n = len(s)
        stack = []
        i = n-1
        
        while i >= 0:
            token = s[i]

            if token.isdigit():
                repeat = ""

                # (1) read the integer of how many times to repeat
                while token.isdigit():
                    repeat = token + repeat
                    i -= 1
                    token = s[i]

                # (2) read the string to be repeated
                stack.pop()
                delta_s = ""
                while stack[-1] != ']':
                    delta_s += stack.pop()
                stack.pop()
                
                # (3) push the repeated string back to stack
                stack.append(int(repeat) * delta_s)
            else:
                stack.append(token)
                i -= 1
        
        stack.reverse()
        return "".join(stack)
````

````python
import heapq
from collections import defaultdict
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # Count the frequency of each element
        numToFreq = defaultdict()
        for num in nums:
            numToFreq[num] = numToFreq[num] + 1

        # Return the top k
        h = []
        nums = list(numToFreq.keys())
        for i in range(k):
            num = nums[i]
            freq = numToFreq[num]
            heapq.heappush(h, (freq, num))
            
        for i in range(k, len(nums)):
            num = nums[i]
            freq = numToFreq[num]
            (kthFreq, kthNum) = heapq.heappop(h)
            print("num", num, "freq", freq,"kthNum", kthNum, "kthFreq", kthFreq, "\n")
            if freq > kthFreq:
                heapq.heappush(h, (freq, num))
            else:
                heapq.heappush(h, (kthFreq, kthNum))
       
        # Collect result
        res = []
        for _ in range(k):
            (freq, num) = heapq.heappop(h)
            res.append(num)
        return res    
````
